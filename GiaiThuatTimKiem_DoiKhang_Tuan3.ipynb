<p>
<b>1.1. Giới thiệu thuật toán đối kháng</b><br>
Thuật toán tìm kiếm đối kháng thực ra đã xuất hiện từ rất sớm, dưới hình thức những ý tưởng chiến lược được áp dụng trong các trò chơi bàn cờ truyền thống như cờ vua hay cờ tướng, vốn luôn phản ánh mối quan hệ cạnh tranh trực tiếp giữa hai người chơi có mục tiêu đối lập. Về mặt khoa học, nền tảng lý thuyết của các thuật toán này bắt nguồn từ lĩnh vực AI cổ điển và đặc biệt là lý thuyết trò chơi (Game Theory), được hình thành từ những năm 1940. Công trình tiên phong của John von Neumann và Oskar Morgenstern đã đặt nền móng với khái niệm trò chơi tổng bằng không (zero-sum games) – mô hình trong đó lợi ích của một người chơi đồng thời là tổn thất của người chơi còn lại.<br>
Cùng thời kỳ đó, thuật toán Minimax ra đời như một bước ngoặt quan trọng. Minimax được đề xuất với mục tiêu giải quyết các trò chơi hai người có tổng bằng không, thông qua nguyên tắc: một người chơi luôn tìm cách tối đa hóa lợi ích của mình trong khi giả định rằng đối thủ sẽ luôn tối thiểu hóa lợi ích đó. Đây được coi là cơ sở lý thuyết vững chắc cho các hệ thống AI chơi trò chơi, giúp máy tính lựa chọn chiến lược tối ưu.<br>
Sau này, với sự ra đời và phát triển của máy tính điện tử, việc triển khai và cải tiến các thuật toán tìm kiếm đối kháng đã trở nên khả thi và mạnh mẽ hơn rất nhiều. Một trong những cải tiến tiêu biểu chính là kỹ thuật cắt tỉa Alpha-Beta (Alpha-Beta Pruning), cho phép giảm đáng kể số trạng thái cần xét trong cây tìm kiếm nhưng vẫn đảm bảo tính đúng đắn của thuật toán Minimax. Nhờ vậy, các hệ thống AI hiện đại có thể xử lý được những trò chơi phức tạp với không gian tìm kiếm khổng lồ như cờ vua hoặc cờ vây.<br>
<b>1.2. Định nghĩa và ý tưởng thuật toán đối kháng</b><br>
Thuật toán tìm kiếm đối kháng (Adversarial Search) là một kỹ thuật cơ bản trong trí tuệ nhân tạo (AI), được sử dụng để giải quyết những bài toán mang tính cạnh tranh giữa hai hoặc nhiều tác nhân có mục tiêu xung đột. Ý tưởng chủ đạo của thuật toán là mô phỏng quá trình ra quyết định trong trò chơi hoặc môi trường đối kháng, nơi mỗi tác nhân đều lựa chọn hành động nhằm tối ưu hóa lợi ích của mình, đồng thời hạn chế khả năng thành công của đối phương. Vai trò của tìm kiếm đối kháng trong AI là mô hình hóa và định hướng hành vi trong các kịch bản ra quyết định cạnh tranh.<br>
Thuật toán này đặc biệt quan trọng ở những lĩnh vực sau: chơi trò chơi (Game Playing), tìm kiếm đối kháng đóng vai trò nền tảng trong việc phát triển AI chơi game. Dù là các trò chơi bàn truyền thống như cờ vua, cờ tướng, cờ vây hay các trò chơi điện tử hiện đại, các tác nhân AI sử dụng tìm kiếm đối kháng để đưa ra những nước đi chiến lược, dự đoán phản ứng của đối phương và tối đa hóa cơ hội chiến thắng trước con người hoặc máy móc.
Hỗ trợ ra quyết định (Decision Making), ngoài lĩnh vực trò chơi, tìm kiếm đối kháng còn có ứng dụng thực tiễn trong nhiều tình huống chiến lược.<br>
•	Kinh tế học: Mô hình hóa các thị trường cạnh tranh, hành vi cạnh tranh, và mối tương tác chiến lược giữa các công ty.<br>
•	Robot học: Hỗ trợ robot tự động xây dựng kế hoạch hành động, có tính tới ý định và hành vi của những đối thủ hoặc tác nhân khác trong cùng môi trường.<br>
Tìm kiếm đối kháng thường phát huy vai trò trong các tình huống cạnh tranh, khi có nhiều tác nhân theo đuổi mục tiêu khác nhau và đối lập. Khi đó:<br>
•	Mỗi tác nhân đều nỗ lực tối đa hóa lợi ích/tiện ích hoặc giảm thiểu tổn thất của bản thân.<br>
•	Hành động của một tác nhân ảnh hưởng trực tiếp tới kết quả cũng như mục tiêu của các tác nhân khác.<br>
•	Các tác nhân không có đầy đủ thông tin về chiến lược của nhau, dẫn tới sự bất định chiến lược.<br>
Trong những bối cảnh như vậy, tìm kiếm đối kháng cung cấp cho hệ thống AI khả năng xác định phương án hành động tối ưu, dự đoán hành vi đối phương và đưa ra phản ứng kịp thời. Nhờ vậy, AI có thể lựa chọn hành động phù hợp nhất nhằm nâng cao xác suất đạt được kết quả thuận lợi.<br>
<b>1.3. Tính chất và thành phần thuật toán</b><br>
Trong tìm kiếm đối kháng, cây trò chơi (Game Tree) là cách biểu diễn trực quan và phổ biến nhất để mô hình hóa quá trình ra quyết định. Cây trò chơi thể hiện lần lượt các hành động và phản ứng có thể xảy ra giữa hai người chơi trong trò chơi cạnh tranh theo từng lượt đi:<br>
•	Mỗi nút (node): Biểu diễn một trạng thái tại một thời điểm nhất định.<br>
•	Mỗi cạnh (edge): Biểu diễn một nước đi hợp lệ chuyển từ trạng thái này sang trạng thái khác.<br>
•	Phân nhánh: Thể hiện tất cả những kịch bản trò chơi có thể xảy ra từ các quyết định khác nhau của người chơi.<br>
Trong cây trò chơi tồn tại hai loại nút đặc biệt: MAX, đại diện cho người chơi tìm cách tối đa hóa điểm số hoặc lợi ích. MIN, đại diện cho người chơi đối phương tìm cách tối thiểu hóa điểm số của MAX. Nguyên tắc cốt lõi là giả định cả hai bên đều chơi tối ưu. Do đó, khi người chơi MAX lựa chọn một nước đi, họ phải cân nhắc dựa trên giả định rằng người chơi MIN sẽ phản ứng theo hướng bất lợi nhất đối với họ (chiến lược xem xét tình huống xấu nhất – Minimax strategy).<br>
Trên thực tế, việc duyệt toàn bộ cây trò chơi thường là bất khả thi với những trò chơi có không gian trạng thái khổng lồ (ví dụ: cờ vua, cờ vây). Vì vậy, các thuật toán như Minimax và đặc biệt là cắt tỉa Alpha–Beta (Alpha–Beta pruning) được sử dụng để điều hướng cây một cách hiệu quả:<br>
•	Minimax: Đánh giá kết quả các trạng thái cuối cùng hoặc trạng thái đến độ sâu xác định rồi lan truyền giá trị lên các nút cha theo nguyên tắc cực tiểu–cực đại.<br>
•	Alpha–Beta pruning: Loại bỏ sớm các nhánh không cần thiết khi đã xác định rằng chúng không ảnh hưởng đến quyết định cuối cùng, từ đó giảm mạnh số nút cần khám phá.<br>
Cây trò chơi không chỉ áp dụng tốt trong các trò chơi kinh điển như cờ vua, cờ tướng, cờ vây, cờ caro, mà còn có thể mô hình hóa nhiều kịch bản cạnh tranh trong đời sống thực như chiến lược thương lượng, đấu thầu hoặc quyết định trong môi trường có đối thủ. Việc sử dụng mô hình cây trò chơi giúp hệ thống AI có thể khảo sát, đánh giá và so sánh nhiều chiến lược khác nhau, từ đó lựa chọn phương án hành động tối ưu nhất.

<b>2.1. Giới thiệu thuật toán MiniMax</b><br>
Thuật toán Minimax là một kỹ thuật cơ bản trong tìm kiếm đối kháng, được thiết kế đặc biệt để đưa ra quyết định tối ưu trong các trò chơi hai người chơi mang tính tổng bằng không. Trong loại trò chơi này, lợi ích của một người chơi luôn đồng nghĩa với thiệt hại của người kia, do đó hai bên có mục tiêu hoàn toàn đối lập. Minimax cho phép người chơi lựa chọn nước đi tối ưu bằng cách giả định rằng đối thủ cũng sẽ luôn lựa chọn nước đi bất lợi nhất đối với mình. Qua đó, người chơi có thể tối đa hóa cơ hội chiến thắng hoặc giảm thiểu nguy cơ thua cuộc.<br>
Người chơi tối đa hóa (Max): Đây là người chơi muốn tối đa hóa lợi ích hoặc điểm số của mình. Trong một ván đấu, người chơi tối đa hóa tìm kiếm những nước đi giúp đạt điểm số cao nhất có thể. Ví dụ, trong cờ vua, người chơi tối đa hóa sẽ muốn thực hiện những nước đi giúp tăng cơ hội chiến thắng.<br>
Thu nhỏ người chơi (Min): Người chơi thu nhỏ nhằm mục đích thu nhỏ tiện ích hoặc điểm số của người chơi tối đa hóa, đóng vai trò là đối thủ, thực hiện các nước đi để chống lại chiến lược của người chơi tối đa hóa và làm giảm cơ hội thành công. Trong cờ vua, người chơi thu nhỏ cố gắng ngăn cản cơ hội chiến thắng của người chơi tối đa hóa.<br>
Giả xử, xét ví dụ minh họa với trò chơi Tic-Tac-Toe, nơi hai người chơi X và O lần lượt đánh dấu trên bảng 3x3.<br>
Trạng thái ban đầu: Bảng 3x3 trống. X là người chơi Max (tối đa hóa), O là người chơi Min (tối thiểu hóa).<br>
Khám phá cây trò chơi: Thuật toán xây dựng cây trò chơi biểu diễn tất cả các trạng thái có thể sau mỗi nước đi.<br>
•	Ở tầng Max: X cố gắng tìm nước đi làm tăng cơ hội chiến thắng.
•	Ở tầng Min: O tìm nước đi làm giảm cơ hội chiến thắng của X.
Đánh giá đệ quy: Thuật toán tiếp tục phân nhánh đến các trạng thái kết thúc (thắng, thua, hòa) hoặc tới độ sâu cho trước.
•	Thắng → giá trị +10.<br>
•	Thua → giá trị -10.<br>
•	Hòa → giá trị 0.<br>
Lan truyền và ra quyết định: Giá trị trạng thái cuối được lan ngược lên cây. Nút Max chọn giá trị lớn nhất từ các con. Nút Min chọn giá trị nhỏ nhất từ các con.<br>
Nước đi chọn lựa: Dựa vào giá trị gốc, thuật toán quyết định nước đi tốt nhất cho X. Tiếp theo trò chơi diễn ra và quá trình được lặp lại.
Bằng cách tuân theo quy trình này, thuật toán Minimax đảm bảo rằng X thực hiện nước đi tốt nhất có thể ở mỗi lượt, tính đến các nước đi phản công của O và tìm kiếm con đường tối ưu để giành chiến thắng.<br>
Tóm lại, thuật toán Minimax là một kỹ thuật then chốt trong tìm kiếm đối kháng, đặc biệt là trong các trò chơi tổng bằng không hai người chơi như Tic-Tac-Toe. Thuật toán sử dụng các khái niệm tối đa hóa và tối thiểu hóa người chơi để xác định nước đi tốt nhất cho mỗi người, dựa trên chiến lược của đối thủ. Điều này cho phép các tác nhân AI đưa ra quyết định chiến lược trong các tình huống cạnh tranh.
