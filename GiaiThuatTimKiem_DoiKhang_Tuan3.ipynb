<b>1.1. Giới thiệu thuật toán đối kháng</b><br>
&emsp;&emsp;Thuật toán tìm kiếm đối kháng thực ra đã xuất hiện từ rất sớm, dưới hình thức những ý tưởng chiến lược được áp dụng trong các trò chơi bàn cờ truyền thống như cờ vua hay cờ tướng, vốn luôn phản ánh mối quan hệ cạnh tranh trực tiếp giữa hai người chơi có mục tiêu đối lập. Về mặt khoa học, nền tảng lý thuyết của các thuật toán này bắt nguồn từ lĩnh vực AI cổ điển và đặc biệt là lý thuyết trò chơi (Game Theory), được hình thành từ những năm 1940. Công trình tiên phong của John von Neumann và Oskar Morgenstern đã đặt nền móng với khái niệm trò chơi tổng bằng không (zero-sum games) – mô hình trong đó lợi ích của một người chơi đồng thời là tổn thất của người chơi còn lại.<br>
&emsp;&emsp;Cùng thời kỳ đó, thuật toán Minimax ra đời như một bước ngoặt quan trọng. Minimax được đề xuất với mục tiêu giải quyết các trò chơi hai người có tổng bằng không, thông qua nguyên tắc: một người chơi luôn tìm cách tối đa hóa lợi ích của mình trong khi giả định rằng đối thủ sẽ luôn tối thiểu hóa lợi ích đó. Đây được coi là cơ sở lý thuyết vững chắc cho các hệ thống AI chơi trò chơi, giúp máy tính lựa chọn chiến lược tối ưu.<br>
&emsp;&emsp;Sau này, với sự ra đời và phát triển của máy tính điện tử, việc triển khai và cải tiến các thuật toán tìm kiếm đối kháng đã trở nên khả thi và mạnh mẽ hơn rất nhiều. Một trong những cải tiến tiêu biểu chính là kỹ thuật cắt tỉa Alpha-Beta (Alpha-Beta Pruning), cho phép giảm đáng kể số trạng thái cần xét trong cây tìm kiếm nhưng vẫn đảm bảo tính đúng đắn của thuật toán Minimax. Nhờ vậy, các hệ thống AI hiện đại có thể xử lý được những trò chơi phức tạp với không gian tìm kiếm khổng lồ như cờ vua hoặc cờ vây.<br>
<b>1.2. Định nghĩa và ý tưởng thuật toán đối kháng</b><br>
&emsp;&emsp;Thuật toán tìm kiếm đối kháng (Adversarial Search) là một kỹ thuật cơ bản trong trí tuệ nhân tạo (AI), được sử dụng để giải quyết những bài toán mang tính cạnh tranh giữa hai hoặc nhiều tác nhân có mục tiêu xung đột. Ý tưởng chủ đạo của thuật toán là mô phỏng quá trình ra quyết định trong trò chơi hoặc môi trường đối kháng, nơi mỗi tác nhân đều lựa chọn hành động nhằm tối ưu hóa lợi ích của mình, đồng thời hạn chế khả năng thành công của đối phương. Vai trò của tìm kiếm đối kháng trong AI là mô hình hóa và định hướng hành vi trong các kịch bản ra quyết định cạnh tranh.<br>
&emsp;&emsp;Thuật toán này đặc biệt quan trọng ở những lĩnh vực sau: chơi trò chơi (Game Playing), tìm kiếm đối kháng đóng vai trò nền tảng trong việc phát triển AI chơi game. Dù là các trò chơi bàn truyền thống như cờ vua, cờ tướng, cờ vây hay các trò chơi điện tử hiện đại, các tác nhân AI sử dụng tìm kiếm đối kháng để đưa ra những nước đi chiến lược, dự đoán phản ứng của đối phương và tối đa hóa cơ hội chiến thắng trước con người hoặc máy móc.Hỗ trợ ra quyết định (Decision Making), ngoài lĩnh vực trò chơi, tìm kiếm đối kháng còn có ứng dụng thực tiễn trong nhiều tình huống chiến lược.<br>
&emsp;&emsp;•	Kinh tế học: Mô hình hóa các thị trường cạnh tranh, hành vi cạnh tranh, và mối tương tác chiến lược giữa các công ty.<br>
&emsp;&emsp;•	Robot học: Hỗ trợ robot tự động xây dựng kế hoạch hành động, có tính tới ý định và hành vi của những đối thủ hoặc tác nhân khác trong cùng môi trường.<br>
&emsp;&emsp;Tìm kiếm đối kháng thường phát huy vai trò trong các tình huống cạnh tranh, khi có nhiều tác nhân theo đuổi mục tiêu khác nhau và đối lập. Khi đó:<br>
&emsp;&emsp;•	Mỗi tác nhân đều nỗ lực tối đa hóa lợi ích/tiện ích hoặc giảm thiểu tổn thất của bản thân.<br>
&emsp;&emsp;•	Hành động của một tác nhân ảnh hưởng trực tiếp tới kết quả cũng như mục tiêu của các tác nhân khác.<br>
&emsp;&emsp;•	Các tác nhân không có đầy đủ thông tin về chiến lược của nhau, dẫn tới sự bất định chiến lược.<br>
&emsp;&emsp;Trong những bối cảnh như vậy, tìm kiếm đối kháng cung cấp cho hệ thống AI khả năng xác định phương án hành động tối ưu, dự đoán hành vi đối phương và đưa ra phản ứng kịp thời. Nhờ vậy, AI có thể lựa chọn hành động phù hợp nhất nhằm nâng cao xác suất đạt được kết quả thuận lợi.<br>
<b>1.3. Tính chất và thành phần thuật toán</b><br>
&emsp;&emsp;Trong tìm kiếm đối kháng, cây trò chơi (Game Tree) là cách biểu diễn trực quan và phổ biến nhất để mô hình hóa quá trình ra quyết định. Cây trò chơi thể hiện lần lượt các hành động và phản ứng có thể xảy ra giữa hai người chơi trong trò chơi cạnh tranh theo từng lượt đi:<br>
&emsp;&emsp;•	Mỗi nút (node): Biểu diễn một trạng thái tại một thời điểm nhất định.<br>
&emsp;&emsp;•	Mỗi cạnh (edge): Biểu diễn một nước đi hợp lệ chuyển từ trạng thái này sang trạng thái khác.<br>
&emsp;&emsp;•	Phân nhánh: Thể hiện tất cả những kịch bản trò chơi có thể xảy ra từ các quyết định khác nhau của người chơi.<br>
&emsp;&emsp;Trong cây trò chơi tồn tại hai loại nút đặc biệt: MAX, đại diện cho người chơi tìm cách tối đa hóa điểm số hoặc lợi ích. MIN, đại diện cho người chơi đối phương tìm cách tối thiểu hóa điểm số của MAX. Nguyên tắc cốt lõi là giả định cả hai bên đều chơi tối ưu. Do đó, khi người chơi MAX lựa chọn một nước đi, họ phải cân nhắc dựa trên giả định rằng người chơi MIN sẽ phản ứng theo hướng bất lợi nhất đối với họ (chiến lược xem xét tình huống xấu nhất – Minimax strategy).<br>
&emsp;&emsp;Trên thực tế, việc duyệt toàn bộ cây trò chơi thường là bất khả thi với những trò chơi có không gian trạng thái khổng lồ (ví dụ: cờ vua, cờ vây). Vì vậy, các thuật toán như Minimax và đặc biệt là cắt tỉa Alpha–Beta (Alpha–Beta pruning) được sử dụng để điều hướng cây một cách hiệu quả:<br>
&emsp;&emsp;•	Minimax: Đánh giá kết quả các trạng thái cuối cùng hoặc trạng thái đến độ sâu xác định rồi lan truyền giá trị lên các nút cha theo nguyên tắc cực tiểu–cực đại.<br>
&emsp;&emsp;•	Alpha–Beta pruning: Loại bỏ sớm các nhánh không cần thiết khi đã xác định rằng chúng không ảnh hưởng đến quyết định cuối cùng, từ đó giảm mạnh số nút cần khám phá.<br>
&emsp;&emsp;Cây trò chơi không chỉ áp dụng tốt trong các trò chơi kinh điển như cờ vua, cờ tướng, cờ vây, cờ caro, mà còn có thể mô hình hóa nhiều kịch bản cạnh tranh trong đời sống thực như chiến lược thương lượng, đấu thầu hoặc quyết định trong môi trường có đối thủ. Việc sử dụng mô hình cây trò chơi giúp hệ thống AI có thể khảo sát, đánh giá và so sánh nhiều chiến lược khác nhau, từ đó lựa chọn phương án hành động tối ưu nhất.

<b>2.1. Cơ sở lý thuyết thuật toán MiniMax</b><br>
<b>2.1.1 Khái niệm thuật toán</b><br>
&emsp;&emsp;Thuật toán Minimax là một kỹ thuật cơ bản trong tìm kiếm đối kháng, được thiết kế đặc biệt để đưa ra quyết định tối ưu trong các trò chơi hai người chơi mang tính tổng bằng không. Trong loại trò chơi này, lợi ích của một người chơi luôn đồng nghĩa với thiệt hại của người kia, do đó hai bên có mục tiêu hoàn toàn đối lập. Minimax cho phép người chơi lựa chọn nước đi tối ưu bằng cách giả định rằng đối thủ cũng sẽ luôn lựa chọn nước đi bất lợi nhất đối với mình. Qua đó, người chơi có thể tối đa hóa cơ hội chiến thắng hoặc giảm thiểu nguy cơ thua cuộc.<br>
&emsp;&emsp;Người chơi tối đa hóa (Max): Đây là người chơi muốn tối đa hóa lợi ích hoặc điểm số của mình. Trong một ván đấu, người chơi tối đa hóa tìm kiếm những nước đi giúp đạt điểm số cao nhất có thể. Ví dụ, trong cờ vua, người chơi tối đa hóa sẽ muốn thực hiện những nước đi giúp tăng cơ hội chiến thắng.<br>
Thu nhỏ người chơi (Min): Người chơi thu nhỏ nhằm mục đích thu nhỏ tiện ích hoặc điểm số của người chơi tối đa hóa, đóng vai trò là đối thủ, thực hiện các nước đi để chống lại chiến lược của người chơi tối đa hóa và làm giảm cơ hội thành công. Trong cờ vua, người chơi thu nhỏ cố gắng ngăn cản cơ hội chiến thắng của người chơi tối đa hóa.<br>
&emsp;&emsp;Giả xử, xét ví dụ minh họa với trò chơi Tic-Tac-Toe, nơi hai người chơi X và O lần lượt đánh dấu trên bảng 3x3. Trạng thái ban đầu, với bảng 3x3 trống, X là người chơi Max (tối đa hóa), O là người chơi Min (tối thiểu hóa). Khám phá cây trò chơi, thuật toán xây dựng cây trò chơi biểu diễn tất cả các trạng thái có thể sau mỗi nước đi.<br>
&emsp;&emsp;•	Ở tầng Max: X cố gắng tìm nước đi làm tăng cơ hội chiến thắng.<br>
&emsp;&emsp;• Ở tầng Min: O tìm nước đi làm giảm cơ hội chiến thắng của X.<br>
&emsp;&emsp;Đánh giá đệ quy: Thuật toán tiếp tục phân nhánh đến các trạng thái kết thúc (thắng, thua, hòa) hoặc tới độ sâu cho trước.<br>
&emsp;&emsp;•	Thắng → giá trị +10.<br>
&emsp;&emsp;•	Thua → giá trị -10.<br>
&emsp;&emsp;•	Hòa → giá trị 0.<br>
&emsp;&emsp;Lan truyền và ra quyết định: Giá trị trạng thái cuối được lan ngược lên cây. Nút Max chọn giá trị lớn nhất từ các con. Nút Min chọn giá trị nhỏ nhất từ các con.<br>
&emsp;&emsp;Nước đi chọn lựa: Dựa vào giá trị gốc, thuật toán quyết định nước đi tốt nhất cho X. Tiếp theo trò chơi diễn ra và quá trình được lặp lại.<br>
Bằng cách tuân theo quy trình này, thuật toán Minimax đảm bảo rằng X thực hiện nước đi tốt nhất có thể ở mỗi lượt, tính đến các nước đi phản công của O và tìm kiếm con đường tối ưu để giành chiến thắng.<br>
&emsp;&emsp;Thuật toán Minimax là một kỹ thuật then chốt trong tìm kiếm đối kháng, đặc biệt là trong các trò chơi tổng bằng không hai người chơi như Tic-Tac-Toe. Thuật toán sử dụng các khái niệm tối đa hóa và tối thiểu hóa người chơi để xác định nước đi tốt nhất cho mỗi người, dựa trên chiến lược của đối thủ. Điều này cho phép các tác nhân AI đưa ra quyết định chiến lược trong các tình huống cạnh tranh.<br>
<b>2.2. Cơ sở lý thuyết thuật toán Alpha-Beta</b><br>
<b>2.2.1. Khái niệm thuật toán</b><br>
&emsp;&emsp;Alpha-Beta là một kỹ thuật tối ưu hóa cho thuật toán Minimax, được sử dụng rộng rãi trong lĩnh vực trí tuệ nhân tạo và lý thuyết trò chơi. Mục tiêu chính của Alpha-Beta là giảm số lượng nút cần đánh giá trong cây trò chơi, từ đó tăng hiệu suất và tốc độ của quá trình ra quyết định.<br>
&emsp;&emsp;Trong thuật toán Minimax truyền thống, tất cả các nút trong cây trò chơi đều được đánh giá để tìm ra nước đi tối ưu. Tuy nhiên, điều này có thể dẫn đến việc xử lý một lượng lớn dữ liệu, đặc biệt trong các trò chơi phức tạp như cờ vua hoặc cờ vây. Alpha-Beta giải quyết vấn đề này bằng cách loại bỏ (prune) các nhánh không cần thiết, dựa trên hai giá trị:<br>
&emsp;&emsp;• Alpha (α): Giá trị tốt nhất mà người chơi tối đa (Maximizer) có thể đảm bảo tại mức đó hoặc cao hơn.<br>
&emsp;&emsp;• Beta (β): Giá trị tốt nhất mà người chơi tối thiểu (Minimizer) có thể đảm bảo tại mức đó hoặc thấp hơn.<br>
&emsp;&emsp;Quá trình hoạt động của Alpha-Beta như sau:<br>
&emsp;&emsp;• Bắt đầu từ gốc của cây trò chơi, thiết lập giá trị ban đầu cho α là âm vô cực và β là dương vô cực.<br>
&emsp;&emsp;• Thực hiện duyệt cây theo chiều sâu (Depth-First Search), cập nhật giá trị α và β tại mỗi nút dựa trên các giá trị con của nó.<br>
&emsp;&emsp;• Nếu tại một nút nào đó, giá trị α lớn hơn hoặc bằng β (α ≥ β), các nhánh con của nút đó sẽ bị cắt tỉa (prune) vì chúng không ảnh hưởng đến quyết định cuối cùng.<br>
&emsp;&emsp;Bằng cách áp dụng Alpha-Beta, thuật toán có thể bỏ qua nhiều nhánh không cần thiết trong cây trò chơi, giúp giảm đáng kể số lượng nút cần đánh giá mà vẫn đảm bảo tìm được nước đi tối ưu. Điều này đặc biệt hữu ích trong các trò chơi có không gian trạng thái lớn, nơi việc đánh giá tất cả các khả năng là không khả thi.<br>
<b>2.4. Các bước thực hiện thuật toán</b><br>
&emsp;&emsp;<b>Bước 1 - Khởi tạo:</b> Bắt đầu từ gốc của cây trò chơi, thiết lập giá trị ban đầu cho α là âm vô cực và β là dương vô cực.<br>
&emsp;&emsp;<b>Bước 2 - Duyệt cây:</b> Thực hiện duyệt cây theo chiều sâu (Depth-First Search), cập nhật giá trị α và β tại mỗi nút dựa trên các giá trị con của nó.<br>
&emsp;&emsp;<b>Bước 3 - Cập nhật giá trị:</b><br>
&emsp;&emsp;Tại mỗi nút, nếu là nút của người chơi tối đa (Maximizer):<br>
&emsp;&emsp;&emsp;&emsp;• Cập nhật α = max(α, giá trị của nút con).<br>
&emsp;&emsp;&emsp;&emsp;• Nếu α ≥ β, dừng việc đánh giá các nút con còn lại (cắt tỉa), vì người chơi tối thiểu sẽ không chọn nhánh này.<br>
&emsp;&emsp;Nếu là nút của người chơi tối thiểu (Minimizer):<br>
&emsp;&emsp;&emsp;&emsp;• Cập nhật β = min(β, giá trị của nút con).<br>
&emsp;&emsp;&emsp;&emsp;• Nếu β ≤ α, dừng việc đánh giá các nút con còn lại (cắt tỉa), vì người chơi tối đa sẽ không chọn nhánh này.<br>
&emsp;&emsp;<b>Bước 4 - Tiếp tục duyệt</b>: Lặp lại quá trình trên cho đến khi duyệt hết các nút cần thiết trong cây.<br>
&emsp;&emsp;Bằng cách áp dụng Alpha-Beta, thuật toán có thể bỏ qua nhiều nhánh không cần thiết trong cây trò chơi, giúp giảm đáng kể số lượng nút cần đánh giá mà vẫn đảm bảo tìm được nước đi tối ưu.
