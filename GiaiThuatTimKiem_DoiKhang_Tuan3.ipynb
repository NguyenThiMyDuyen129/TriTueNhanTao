<p><b>1.1. Giới thiệu thuật toán đối kháng</b></p>
<p>
Thuật toán tìm kiếm đối kháng thực ra đã xuất hiện từ rất sớm, dưới hình thức những ý tưởng chiến lược được áp dụng trong các trò chơi bàn cờ truyền thống như cờ vua hay cờ tướng, vốn luôn phản ánh mối quan hệ cạnh tranh trực tiếp giữa hai người chơi có mục tiêu đối lập.
Về mặt khoa học, nền tảng lý thuyết của các thuật toán này bắt nguồn từ lĩnh vực AI cổ điển và đặc biệt là lý thuyết trò chơi (Game Theory), được hình thành từ những năm 1940. Công trình tiên phong của John von Neumann và Oskar Morgenstern đã đặt nền móng với khái niệm trò chơi tổng bằng không (zero-sum games) – mô hình trong đó lợi ích của một người chơi đồng thời là tổn thất của người chơi còn lại.

Cùng thời kỳ đó, thuật toán Minimax ra đời như một bước ngoặt quan trọng. Minimax được đề xuất với mục tiêu giải quyết các trò chơi hai người có tổng bằng không, thông qua nguyên tắc: một người chơi luôn tìm cách tối đa hóa lợi ích của mình trong khi giả định rằng đối thủ sẽ luôn tối thiểu hóa lợi ích đó. Đây được coi là cơ sở lý thuyết vững chắc cho các hệ thống AI chơi trò chơi, giúp máy tính lựa chọn chiến lược tối ưu.

Sau này, với sự ra đời và phát triển của máy tính điện tử, việc triển khai và cải tiến các thuật toán tìm kiếm đối kháng đã trở nên khả thi và mạnh mẽ hơn rất nhiều. Một trong những cải tiến tiêu biểu chính là kỹ thuật cắt tỉa Alpha-Beta (Alpha-Beta Pruning), cho phép giảm đáng kể số trạng thái cần xét trong cây tìm kiếm nhưng vẫn đảm bảo tính đúng đắn của thuật toán Minimax. Nhờ vậy, các hệ thống AI hiện đại có thể xử lý được những trò chơi phức tạp với không gian tìm kiếm khổng lồ như cờ vua hoặc cờ vây.
