<b>1.1. Giới thiệu thuật toán đối kháng</b><br>
&emsp;&emsp;Thuật toán tìm kiếm đối kháng thực ra đã xuất hiện từ rất sớm, dưới hình thức những ý tưởng chiến lược được áp dụng trong các trò chơi bàn cờ truyền thống như cờ vua hay cờ tướng, vốn luôn phản ánh mối quan hệ cạnh tranh trực tiếp giữa hai người chơi có mục tiêu đối lập. Về mặt khoa học, nền tảng lý thuyết của các thuật toán này bắt nguồn từ lĩnh vực AI cổ điển và đặc biệt là lý thuyết trò chơi (Game Theory), được hình thành từ những năm 1940. Công trình tiên phong của John von Neumann và Oskar Morgenstern đã đặt nền móng với khái niệm trò chơi tổng bằng không (zero-sum games) – mô hình trong đó lợi ích của một người chơi đồng thời là tổn thất của người chơi còn lại.<br>
&emsp;&emsp;Cùng thời kỳ đó, thuật toán Minimax ra đời như một bước ngoặt quan trọng. Minimax được đề xuất với mục tiêu giải quyết các trò chơi hai người có tổng bằng không, thông qua nguyên tắc: một người chơi luôn tìm cách tối đa hóa lợi ích của mình trong khi giả định rằng đối thủ sẽ luôn tối thiểu hóa lợi ích đó. Đây được coi là cơ sở lý thuyết vững chắc cho các hệ thống AI chơi trò chơi, giúp máy tính lựa chọn chiến lược tối ưu.<br>
&emsp;&emsp;Sau này, với sự ra đời và phát triển của máy tính điện tử, việc triển khai và cải tiến các thuật toán tìm kiếm đối kháng đã trở nên khả thi và mạnh mẽ hơn rất nhiều. Một trong những cải tiến tiêu biểu chính là kỹ thuật cắt tỉa Alpha-Beta (Alpha-Beta Pruning), cho phép giảm đáng kể số trạng thái cần xét trong cây tìm kiếm nhưng vẫn đảm bảo tính đúng đắn của thuật toán Minimax. Nhờ vậy, các hệ thống AI hiện đại có thể xử lý được những trò chơi phức tạp với không gian tìm kiếm khổng lồ như cờ vua hoặc cờ vây.<br>
<b>1.2. Định nghĩa và ý tưởng thuật toán đối kháng</b><br>
&emsp;&emsp;Thuật toán tìm kiếm đối kháng (Adversarial Search) là một kỹ thuật cơ bản trong trí tuệ nhân tạo (AI), được sử dụng để giải quyết những bài toán mang tính cạnh tranh giữa hai hoặc nhiều tác nhân có mục tiêu xung đột. Ý tưởng chủ đạo của thuật toán là mô phỏng quá trình ra quyết định trong trò chơi hoặc môi trường đối kháng, nơi mỗi tác nhân đều lựa chọn hành động nhằm tối ưu hóa lợi ích của mình, đồng thời hạn chế khả năng thành công của đối phương. Vai trò của tìm kiếm đối kháng trong AI là mô hình hóa và định hướng hành vi trong các kịch bản ra quyết định cạnh tranh.<br>
&emsp;&emsp;Thuật toán này đặc biệt quan trọng ở những lĩnh vực sau: chơi trò chơi (Game Playing), tìm kiếm đối kháng đóng vai trò nền tảng trong việc phát triển AI chơi game. Dù là các trò chơi bàn truyền thống như cờ vua, cờ tướng, cờ vây hay các trò chơi điện tử hiện đại, các tác nhân AI sử dụng tìm kiếm đối kháng để đưa ra những nước đi chiến lược, dự đoán phản ứng của đối phương và tối đa hóa cơ hội chiến thắng trước con người hoặc máy móc.Hỗ trợ ra quyết định (Decision Making), ngoài lĩnh vực trò chơi, tìm kiếm đối kháng còn có ứng dụng thực tiễn trong nhiều tình huống chiến lược.<br>
&emsp;&emsp;•	Kinh tế học: Mô hình hóa các thị trường cạnh tranh, hành vi cạnh tranh, và mối tương tác chiến lược giữa các công ty.<br>
&emsp;&emsp;•	Robot học: Hỗ trợ robot tự động xây dựng kế hoạch hành động, có tính tới ý định và hành vi của những đối thủ hoặc tác nhân khác trong cùng môi trường.<br>
&emsp;&emsp;Tìm kiếm đối kháng thường phát huy vai trò trong các tình huống cạnh tranh, khi có nhiều tác nhân theo đuổi mục tiêu khác nhau và đối lập. Khi đó:<br>
&emsp;&emsp;•	Mỗi tác nhân đều nỗ lực tối đa hóa lợi ích/tiện ích hoặc giảm thiểu tổn thất của bản thân.<br>
&emsp;&emsp;•	Hành động của một tác nhân ảnh hưởng trực tiếp tới kết quả cũng như mục tiêu của các tác nhân khác.<br>
&emsp;&emsp;•	Các tác nhân không có đầy đủ thông tin về chiến lược của nhau, dẫn tới sự bất định chiến lược.<br>
&emsp;&emsp;Trong những bối cảnh như vậy, tìm kiếm đối kháng cung cấp cho hệ thống AI khả năng xác định phương án hành động tối ưu, dự đoán hành vi đối phương và đưa ra phản ứng kịp thời. Nhờ vậy, AI có thể lựa chọn hành động phù hợp nhất nhằm nâng cao xác suất đạt được kết quả thuận lợi.<br>
<b>1.3. Tính chất và thành phần thuật toán</b><br>
&emsp;&emsp;Trong tìm kiếm đối kháng, cây trò chơi (Game Tree) là cách biểu diễn trực quan và phổ biến nhất để mô hình hóa quá trình ra quyết định. Cây trò chơi thể hiện lần lượt các hành động và phản ứng có thể xảy ra giữa hai người chơi trong trò chơi cạnh tranh theo từng lượt đi:<br>
&emsp;&emsp;•	Mỗi nút (node): Biểu diễn một trạng thái tại một thời điểm nhất định.<br>
&emsp;&emsp;•	Mỗi cạnh (edge): Biểu diễn một nước đi hợp lệ chuyển từ trạng thái này sang trạng thái khác.<br>
&emsp;&emsp;•	Phân nhánh: Thể hiện tất cả những kịch bản trò chơi có thể xảy ra từ các quyết định khác nhau của người chơi.<br>
&emsp;&emsp;Trong cây trò chơi tồn tại hai loại nút đặc biệt: MAX, đại diện cho người chơi tìm cách tối đa hóa điểm số hoặc lợi ích. MIN, đại diện cho người chơi đối phương tìm cách tối thiểu hóa điểm số của MAX. Nguyên tắc cốt lõi là giả định cả hai bên đều chơi tối ưu. Do đó, khi người chơi MAX lựa chọn một nước đi, họ phải cân nhắc dựa trên giả định rằng người chơi MIN sẽ phản ứng theo hướng bất lợi nhất đối với họ (chiến lược xem xét tình huống xấu nhất – Minimax strategy).<br>
&emsp;&emsp;Trên thực tế, việc duyệt toàn bộ cây trò chơi thường là bất khả thi với những trò chơi có không gian trạng thái khổng lồ (ví dụ: cờ vua, cờ vây). Vì vậy, các thuật toán như Minimax và đặc biệt là cắt tỉa Alpha–Beta (Alpha–Beta pruning) được sử dụng để điều hướng cây một cách hiệu quả:<br>
&emsp;&emsp;•	Minimax: Đánh giá kết quả các trạng thái cuối cùng hoặc trạng thái đến độ sâu xác định rồi lan truyền giá trị lên các nút cha theo nguyên tắc cực tiểu–cực đại.<br>
&emsp;&emsp;•	Alpha–Beta pruning: Loại bỏ sớm các nhánh không cần thiết khi đã xác định rằng chúng không ảnh hưởng đến quyết định cuối cùng, từ đó giảm mạnh số nút cần khám phá.<br>
&emsp;&emsp;Cây trò chơi không chỉ áp dụng tốt trong các trò chơi kinh điển như cờ vua, cờ tướng, cờ vây, cờ caro, mà còn có thể mô hình hóa nhiều kịch bản cạnh tranh trong đời sống thực như chiến lược thương lượng, đấu thầu hoặc quyết định trong môi trường có đối thủ. Việc sử dụng mô hình cây trò chơi giúp hệ thống AI có thể khảo sát, đánh giá và so sánh nhiều chiến lược khác nhau, từ đó lựa chọn phương án hành động tối ưu nhất.

<b>2.1. Cơ sở lý thuyết thuật toán MiniMax</b><br>
<b>2.1.1 Khái niệm thuật toán</b><br>
&emsp;&emsp;Thuật toán Minimax là một kỹ thuật cơ bản trong tìm kiếm đối kháng, được thiết kế đặc biệt để đưa ra quyết định tối ưu trong các trò chơi hai người chơi mang tính tổng bằng không. Trong loại trò chơi này, lợi ích của một người chơi luôn đồng nghĩa với thiệt hại của người kia, do đó hai bên có mục tiêu hoàn toàn đối lập. Minimax cho phép người chơi lựa chọn nước đi tối ưu bằng cách giả định rằng đối thủ cũng sẽ luôn lựa chọn nước đi bất lợi nhất đối với mình. Qua đó, người chơi có thể tối đa hóa cơ hội chiến thắng hoặc giảm thiểu nguy cơ thua cuộc.<br>
&emsp;&emsp;Người chơi tối đa hóa (Max): Đây là người chơi muốn tối đa hóa lợi ích hoặc điểm số của mình. Trong một ván đấu, người chơi tối đa hóa tìm kiếm những nước đi giúp đạt điểm số cao nhất có thể. Ví dụ, trong cờ vua, người chơi tối đa hóa sẽ muốn thực hiện những nước đi giúp tăng cơ hội chiến thắng.<br>
Thu nhỏ người chơi (Min): Người chơi thu nhỏ nhằm mục đích thu nhỏ tiện ích hoặc điểm số của người chơi tối đa hóa, đóng vai trò là đối thủ, thực hiện các nước đi để chống lại chiến lược của người chơi tối đa hóa và làm giảm cơ hội thành công. Trong cờ vua, người chơi thu nhỏ cố gắng ngăn cản cơ hội chiến thắng của người chơi tối đa hóa.<br>
&emsp;&emsp;Giả xử, xét ví dụ minh họa với trò chơi Tic-Tac-Toe, nơi hai người chơi X và O lần lượt đánh dấu trên bảng 3x3. Trạng thái ban đầu, với bảng 3x3 trống, X là người chơi Max (tối đa hóa), O là người chơi Min (tối thiểu hóa). Khám phá cây trò chơi, thuật toán xây dựng cây trò chơi biểu diễn tất cả các trạng thái có thể sau mỗi nước đi.<br>
&emsp;&emsp;•	Ở tầng Max: X cố gắng tìm nước đi làm tăng cơ hội chiến thắng.<br>
&emsp;&emsp;• Ở tầng Min: O tìm nước đi làm giảm cơ hội chiến thắng của X.<br>
&emsp;&emsp;Đánh giá đệ quy: Thuật toán tiếp tục phân nhánh đến các trạng thái kết thúc (thắng, thua, hòa) hoặc tới độ sâu cho trước.<br>
&emsp;&emsp;•	Thắng → giá trị +10.<br>
&emsp;&emsp;•	Thua → giá trị -10.<br>
&emsp;&emsp;•	Hòa → giá trị 0.<br>
&emsp;&emsp;Lan truyền và ra quyết định: Giá trị trạng thái cuối được lan ngược lên cây. Nút Max chọn giá trị lớn nhất từ các con. Nút Min chọn giá trị nhỏ nhất từ các con.<br>
&emsp;&emsp;Nước đi chọn lựa: Dựa vào giá trị gốc, thuật toán quyết định nước đi tốt nhất cho X. Tiếp theo trò chơi diễn ra và quá trình được lặp lại.<br>
Bằng cách tuân theo quy trình này, thuật toán Minimax đảm bảo rằng X thực hiện nước đi tốt nhất có thể ở mỗi lượt, tính đến các nước đi phản công của O và tìm kiếm con đường tối ưu để giành chiến thắng.<br>
&emsp;&emsp;Thuật toán Minimax là một kỹ thuật then chốt trong tìm kiếm đối kháng, đặc biệt là trong các trò chơi tổng bằng không hai người chơi như Tic-Tac-Toe. Thuật toán sử dụng các khái niệm tối đa hóa và tối thiểu hóa người chơi để xác định nước đi tốt nhất cho mỗi người, dựa trên chiến lược của đối thủ. Điều này cho phép các tác nhân AI đưa ra quyết định chiến lược trong các tình huống cạnh tranh.<br>
<b>2.1.2. Quy trình thực hiện</b><br>
&emsp;&emsp;Thuật toán Minimax hoạt động theo cơ chế Tìm kiếm theo chiều sâu (Depth-First Search - DFS) và Lan truyền ngược (Back-propagation). Quy trình cụ thể gồm 4 bước:<br>
&emsp;&emsp;1. Sinh cây trò chơi: Từ trạng thái hiện tại, thuật toán sinh ra các trạng thái con cho đến độ sâu quy định hoặc đến trạng thái kết thúc.<br>
&emsp;&emsp;2. Đánh giá tại lá: Tại các nút lá (leaf nodes), áp dụng hàm lợi ích (hoặc hàm Heuristic) để tính điểm số cho trạng thái đó.<br>
&emsp;&emsp;3. Lan truyền ngược (Backtracking): Giá trị từ các nút lá được đẩy ngược lên các nút cha theo quy tắc:<br>
&emsp;&emsp;&emsp;&emsp;+ Tại tầng của MIN, nút cha nhận giá trị nhỏ nhất của các con.<br>
&emsp;&emsp;&emsp;&emsp;+ Tại tầng của MAX, nút cha nhận giá trị lớn nhất của các con.<br>
&emsp;&emsp;4. Ra quyết định: Khi giá trị được lan truyền ngược về đến nút gốc (Root), MAX sẽ chọn nước đi dẫn đến nút con có giá trị cao nhất.<br>
<b>2.2. Cơ sở lý thuyết thuật toán Alpha-Beta</b><br>
<b>2.2.1. Khái niệm thuật toán</b><br>
&emsp;&emsp;Alpha-Beta là một kỹ thuật tối ưu hóa cho thuật toán Minimax, được sử dụng rộng rãi trong lĩnh vực trí tuệ nhân tạo và lý thuyết trò chơi. Mục tiêu chính của Alpha-Beta là giảm số lượng nút cần đánh giá trong cây trò chơi, từ đó tăng hiệu suất và tốc độ của quá trình ra quyết định.<br>
&emsp;&emsp;Trong thuật toán Minimax truyền thống, tất cả các nút trong cây trò chơi đều được đánh giá để tìm ra nước đi tối ưu. Tuy nhiên, điều này có thể dẫn đến việc xử lý một lượng lớn dữ liệu, đặc biệt trong các trò chơi phức tạp như cờ vua hoặc cờ vây. Alpha-Beta giải quyết vấn đề này bằng cách loại bỏ (prune) các nhánh không cần thiết, dựa trên hai giá trị:<br>
&emsp;&emsp;• Alpha (α): Giá trị tốt nhất mà người chơi tối đa (Maximizer) có thể đảm bảo tại mức đó hoặc cao hơn.<br>
&emsp;&emsp;• Beta (β): Giá trị tốt nhất mà người chơi tối thiểu (Minimizer) có thể đảm bảo tại mức đó hoặc thấp hơn.<br>
&emsp;&emsp;Quá trình hoạt động của Alpha-Beta như sau:<br>
&emsp;&emsp;• Bắt đầu từ gốc của cây trò chơi, thiết lập giá trị ban đầu cho α là âm vô cực và β là dương vô cực.<br>
&emsp;&emsp;• Thực hiện duyệt cây theo chiều sâu (Depth-First Search), cập nhật giá trị α và β tại mỗi nút dựa trên các giá trị con của nó.<br>
&emsp;&emsp;• Nếu tại một nút nào đó, giá trị α lớn hơn hoặc bằng β (α ≥ β), các nhánh con của nút đó sẽ bị cắt tỉa (prune) vì chúng không ảnh hưởng đến quyết định cuối cùng.<br>
&emsp;&emsp;Bằng cách áp dụng Alpha-Beta, thuật toán có thể bỏ qua nhiều nhánh không cần thiết trong cây trò chơi, giúp giảm đáng kể số lượng nút cần đánh giá mà vẫn đảm bảo tìm được nước đi tối ưu. Điều này đặc biệt hữu ích trong các trò chơi có không gian trạng thái lớn, nơi việc đánh giá tất cả các khả năng là không khả thi.<br>
<b>2.4. Các bước thực hiện thuật toán</b><br>
&emsp;&emsp;<b>Bước 1 - Khởi tạo:</b> Bắt đầu từ gốc của cây trò chơi, thiết lập giá trị ban đầu cho α là âm vô cực và β là dương vô cực.<br>
&emsp;&emsp;<b>Bước 2 - Duyệt cây:</b> Thực hiện duyệt cây theo chiều sâu (Depth-First Search), cập nhật giá trị α và β tại mỗi nút dựa trên các giá trị con của nó.<br>
&emsp;&emsp;<b>Bước 3 - Cập nhật giá trị:</b><br>
&emsp;&emsp;Tại mỗi nút, nếu là nút của người chơi tối đa (Maximizer):<br>
&emsp;&emsp;&emsp;&emsp;• Cập nhật α = max(α, giá trị của nút con).<br>
&emsp;&emsp;&emsp;&emsp;• Nếu α ≥ β, dừng việc đánh giá các nút con còn lại (cắt tỉa), vì người chơi tối thiểu sẽ không chọn nhánh này.<br>
&emsp;&emsp;Nếu là nút của người chơi tối thiểu (Minimizer):<br>
&emsp;&emsp;&emsp;&emsp;• Cập nhật β = min(β, giá trị của nút con).<br>
&emsp;&emsp;&emsp;&emsp;• Nếu β ≤ α, dừng việc đánh giá các nút con còn lại (cắt tỉa), vì người chơi tối đa sẽ không chọn nhánh này.<br>
&emsp;&emsp;<b>Bước 4 - Tiếp tục duyệt</b>: Lặp lại quá trình trên cho đến khi duyệt hết các nút cần thiết trong cây.<br>

Thuật toán này mô phỏng quá trình ra quyết định bằng cách duyệt toàn bộ cây trạng thái của trò chơi, đánh giá tất cả các nước đi có thể xảy ra và lựa chọn phương án tốt nhất cho người chơi máy (AI). Để triển khai một cách hiệu quả, chương trình sử dụng một số thư viện Python hỗ trợ quá trình xử lý:<br>

&emsp;&emsp;Bên cạnh đó, chương trình sử dụng các hằng số như X, O và EMPTY để biểu diễn hai người chơi và các ô trống trên bàn cờ. Đồng thời quản lý lượt chơi giữa người dùng và máy thông qua các biến user và ai. Sự kết hợp giữa thuật toán Minimax và các thư viện hỗ trợ giúp hệ thống hoạt động chính xác, hiệu quả và đưa ra chiến lược tối ưu trong mọi tình huống của trò chơi.

```python
import copy
import math
import random
import numpy

X = "X"
O = "O"
EMPTY = None
user = None
ai = None
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm initial_state</i><br>
&emsp;&emsp;Hàm initial_state() được sử dụng để khởi tạo trạng thái ban đầu của bàn cờ trong trò chơi Tic-Tac-Toe. Đây là điểm xuất phát cho toàn bộ thuật toán Minimax cũng như quá trình người chơi và máy tính bắt đầu tương tác. Cụ thể, hàm trả về một cấu trúc dữ liệu dạng ma trận (3×3), trong đó mỗi ô mang giá trị EMPTY để thể hiện rằng chưa có người chơi nào đánh vào vị trí đó.<br>
</div>
</font>

```python
def initial_state():                      #Hàm không nhận tham số và có nhiệm vụ tạo ra một bàn cờ mới hoàn toàn.
    return [[EMPTY, EMPTY, EMPTY],        #Hàm trả về trạng thái khởi đầu của bàn cờ.
            [EMPTY, EMPTY, EMPTY],        #Đây là ma trận 3×3 mô phỏng bàn cờ Tic-Tac-Toe.
            [EMPTY, EMPTY, EMPTY]]        #Mỗi phần tử nhận giá trị EMPTY (được định nghĩa trước đó là None), biểu thị ô trống.
            #Bàn cờ được tổ chức dạng danh sách lồng danh sách (list of lists), thuận tiện cho việc:
            #+ kiểm tra dòng, cột, đường chéo;
            #+ xác định nước đi hợp lệ;
            #+ tính toán trạng thái tiếp theo trong Minimax.
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm player(board)</i><br>
&emsp;&emsp;Hàm player() có nhiệm vụ xác định người chơi nào sẽ đi tiếp theo trên bàn cờ hiện tại. Đây là một thành phần quan trọng trong thuật toán Minimax, giúp xác định bên nào đang cần đánh — từ đó chọn chiến lược MAX hoặc MIN phù hợp.<br>
</div>
</font>

```python
def player(board):                        #Hàm nhận vào trạng thái hiện tại của bàn cờ.
    count = 0 #Đếm số ô đã được đánh      #Hàm trả về người chơi sẽ đánh nước tiếp theo.
    for i in board:
      for j in i:
        if j:
          count += 1
          if count % 2 != 0:
            return ai
    return user
    #Mỗi khi một ô có giá trị khác EMPTY (tức có X hoặc O), biến count tăng 1.
    #Tổng count cho biết đã có bao nhiêu nước được đánh từ đầu ván đến hiện tại.
    #+ Nếu số ô đã đánh là số lẻ → nghĩa là người dùng (user) đi trước → tới lượt AI.
    #+ Nếu số ô đã đánh là số chẵn → tới lượt người dùng.
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm actions(board)</i><br>
&emsp;&emsp;Hàm actions() có nhiệm vụ xác định tất cả nước đi hợp lệ trên bàn cờ hiện tại. Những nước đi này chính là các nhánh của cây trạng thái mà thuật toán Minimax sẽ duyệt. Hàm này đóng vai trò quan trọng trong việc sinh trạng thái mới (result) và mở rộng toàn bộ không gian tìm kiếm của Minimax.<br>
</div>
</font>

```python
def actions(board):
    #Hàm trả về tập hợp tất cả các ô mà người chơi có thể đánh tiếp.
    res = set() #lưu các vị trí trống.
    board_len = len(board)
    for i in range(board_len):#Hai vòng lặp duyệt từng hàng và từng cột.
      for j in range(board_len):
        if board[i][j] == EMPTY:#Nếu ô (i, j) có giá trị EMPTY → đây là một nước đi hợp lệ.
          res.add((i, j))#Thêm vào tập res dưới dạng cặp tọa độ (i, j).
    return res #Một tập các nước đi:{(0, 0), (0, 1), (1, 2), …}. Nếu bàn cờ đầy → trả về {} (tập rỗng)
    #Chỉ những ô chưa có X hoặc O mới được phép đánh.
    #Đây là danh sách dùng để:
    #+ Sinh trạng thái kế tiếp cho Minimax,
    #+ Kiểm tra xem trò chơi còn tiếp tục hay đã hòa.
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm result(board, action)</i><br>
&emsp;&emsp;Hàm result() mô phỏng việc thực hiện một nước đi trên bàn cờ. Đây là hàm cốt lõi của thuật toán Minimax, giúp sinh ra trạng thái mới (state) từ trạng thái hiện tại. Mỗi action tạo ra một node con trong cây trạng thái mà Minimax sẽ đệ quy duyệt.<br>
</div>
</font>

```python
def result(board, action):                  #board: trạng thái hiện tại, action: một tuple (i, j) đại diện cho ô mà người chơi muốn đánh.
                                            #trả về một bàn cờ mới, là kết quả sau khi thực hiện nước đi.
    curr_player = player(board)             #Gọi hàm player() để biết ai đang được phép đánh (user hay AI).
    result_board = copy.deepcopy(board)     #Tránh làm thay đổi trạng thái board ban đầu vì mỗi state là một cây độc lập. Nếu dùng bản sao nông (shallow copy), các thay đổi sẽ ảnh hưởng lẫn nhau → sai toàn bộ thuật toán.
    (i, j) = action                         #Lấy tọa độ hàng và cột mà người chơi muốn đánh.
    result_board[i][j] = curr_player        #Thực hiện nước đi bằng cách đặt ký hiệu: "X" hoặc "O". Tùy theo lượt được trả về bởi player(board).
    return result_board                     #Đây là node con của cây Minimax.
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm get_horizontal_winner(board)</i><br>
&emsp;&emsp;Hàm result() mô phỏng việc thực hiện một nước đi trên bàn cờ. Đây là hàm cốt lõi của thuật toán Minimax, giúp sinh ra trạng thái mới (state) từ trạng thái hiện tại. Mỗi action tạo ra một node con trong cây trạng thái mà Minimax sẽ đệ quy duyệt.<br>
</div>
</font>

```python
def get_horizontal_winner(board):                   #Hàm nhận vào bàn cờ 3×3 và kiểm tra từng hàng.
    # hàm kiểm tra theo chiều ngang.
    winner_val = None                               #lưu ký hiệu của người chơi đang được kiểm tra (X hoặc O).
    board_len = len(board)                          #số hàng của bàn cờ (thường là 3).
    for i in range(board_len):
        winner_val = board[i][0]                    #Lấy giá trị của ô đầu tiên trong hàng i làm chuẩn. Nếu toàn bộ hàng giống với ô này → người chơi đó thắng.
        for j in range(board_len):                  #Duyệt từng ô trong hàng i. Nếu bất kỳ ô nào khác ký hiệu chuẩn (winner_val) → hàng này không thắng. Đặt lại winner_val thành None để báo không có người thắng ở hàng này.
            if board[i][j] != winner_val:           #Nếu không có ô nào bị loại → winner_val vẫn giữ là "X" hoặc "O".
                winner_val = None                   #Như vậy, người đó là người thắng cuối cùng theo hàng ngang.
                                                    #Nếu duyệt xong mọi hàng mà không tìm được hàng thắng → trả về None.
        if winner_val:
            return winner_val
    return winner_val
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm get_vertical_winner(board)</i><br>
&emsp;&emsp;Hàm get_vertical_winner() kiểm tra xem có người chơi nào chiến thắng theo chiều dọc (cột) hay không. Đây là một trong ba phương thức xác định chiến thắng trong trò chơi TicTacToe, bên cạnh việc kiểm tra hàng ngang và đường chéo.<br>
&emsp;&emsp;Thông tin về người thắng là yếu tố quan trọng trong thuật toán Minimax vì nó giúp nhận biết trạng thái kết thúc và tính toán điểm utility() để lan ngược lên cây trạng thái.<br>
</div>
</font>

```python
def get_vertical_winner(board):                       #Hàm nhận vào bàn cờ 3×3 và kiểm tra từng cột để xem có người thắng theo chiều dọc hay không.
    # check vertically                                #Hàm kiểm tra theo chiều dọc.
    winner_val = None                                 #Lưu ký hiệu ban đầu của người chơi đang được kiểm tra (X hoặc O).
    board_len = len(board)                            #Số cột của bàn cờ (thường là 3 với Tic-Tac-Toe).
    for i in range(board_len):                        #Duyệt qua từng cột trên bàn cờ.
        winner_val = board[0][i]                      #Lấy giá trị của ô đầu tiên trong cột i làm giá trị chuẩn. Nếu toàn bộ cột mang giá trị giống nhau → người chơi đó thắng.
        for j in range(board_len):                    #Duyệt từng ô trong cột i. Nếu ô nào khác giá trị chuẩn → cột này không có người thắng.
            if board[j][i] != winner_val:             #So sánh từng ô trong cột với winner_val.
                winner_val = None                     #Nếu thấy ô khác ⇒ đặt winner_val về None để báo cột này không hợp lệ.
                                                      #Nếu duyệt hết mà không bị loại, winner_val vẫn giữ "X" hoặc "O".
        if winner_val:                                #Nếu winner_val còn giữ giá trị hợp lệ ⇒ đây là người chiến thắng theo chiều dọc.
            return winner_val                         #Trả về ký hiệu người thắng.
    return winner_val                                 #Nếu duyệt hết tất cả cột mà không có ai thắng → trả về None.
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm get_diagonal_winner(board)</i><br>
&emsp;&emsp;Hàm get_diagonal_winner(board) có nhiệm vụ kiểm tra xem bàn cờ có người thắng theo đường chéo hay không. Trong Tic-Tac-Toe, có hai đường chéo cần được kiểm tra: đường chéo chính (từ góc trái trên xuống góc phải dưới) và đường chéo phụ (từ góc phải trên xuống góc trái dưới). Hàm sẽ so sánh các ô trong từng đường chéo với ô đầu tiên làm chuẩn. Nếu tất cả đều giống nhau và không phải là ô trống, người chơi đó là người chiến thắng. Đây là phần quan trọng trong quá trình xác định trạng thái kết thúc của trò chơi, giúp thuật toán Minimax biết khi nào phải dừng mở rộng cây tìm kiếm và trả về giá trị đánh giá tương ứng.<br>
</div>
</font>

```python
def get_diagonal_winner(board):                        #Hàm nhận vào bàn cờ và kiểm tra hai đường chéo để xác định người thắng.
    # check diagonally                                   #Hàm kiểm tra chiến thắng theo đường chéo.
    winner_val = None                                   #Biến lưu ký hiệu của người chơi đang được kiểm tra (X hoặc O).
    board_len = len(board)                              #Độ dài cạnh bàn cờ (3 với Tic-Tac-Toe).

    winner_val = board[0][0]                            #Lấy ô đầu tiên của đường chéo chính (0,0) làm chuẩn.
    for i in range(board_len):                          #Duyệt từng ô trên đường chéo chính.
        if board[i][i] != winner_val:                   #Nếu bất kỳ ô nào khác chuẩn → không có người thắng ở đường chéo chính.
            winner_val = None                           #Đặt lại winner_val nếu không khớp.
    if winner_val:                                      #Nếu tất cả ô đều giống nhau → người đó thắng.
        return winner_val                               #Trả về người thắng ở đường chéo chính.

    winner_val = board[0][board_len - 1]                #Lấy ô đầu tiên của đường chéo phụ (0,2) làm chuẩn.
    for i in range(board_len):                          #Duyệt các ô tạo thành đường chéo phụ.
        j = board_len - 1 - i                           #Cột giảm dần từ phải sang trái.
        if board[i][j] != winner_val:                   #Nếu ô nào khác giá trị chuẩn → không có người thắng đường chéo phụ.
            winner_val = None
    return winner_val                                   #Nếu có người thắng → trả về X hoặc O, ngược lại trả về None.
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm winner(board)</i><br>
&emsp;&emsp;Hàm winner(board) giữ vai trò tổng hợp kết quả thắng từ ba phương thức kiểm tra: hàng ngang, hàng dọc và đường chéo. Thay vì kiểm tra trực tiếp trong hàm chính, hàm này gọi tuần tự các hàm get_horizontal_winner, get_vertical_winner và get_diagonal_winner để xác định người chơi chiến thắng. Nếu bất kỳ phương thức nào phát hiện có người thắng, ký hiệu của người đó sẽ được trả về ngay. Nếu cả ba phương thức đều không cho kết quả, hàm trả về None. Đây là một thành phần then chốt để thuật toán Minimax biết khi nào trò chơi kết thúc, từ đó quyết định điểm đánh giá (utility) của trạng thái hiện tại. Nhờ hàm này, việc phát hiện chiến thắng trở nên rõ ràng, tách biệt và dễ bảo trì.<br>
</div>
</font>

```python
def winner(board):                                        #Hàm xác định xem có người chơi nào thắng trên bàn cờ hay không.
    """
    Returns the winner of the game, if there is one.
    """
    winner_val = (
        get_horizontal_winner(board)                       #Gọi hàm kiểm tra thắng theo hàng ngang. Nếu có người thắng, giá trị trả về sẽ là "X" hoặc "O".
        or get_vertical_winner(board)                      #Nếu chưa tìm thấy, tiếp tục kiểm tra thắng theo cột.
        or get_diagonal_winner(board)                      #Nếu vẫn chưa có kết quả, kiểm tra hai đường chéo.
        or None                                            #Nếu cả 3 phương thức đều trả về None → không có người thắng.
    )
    return winner_val                                      #Kết quả cuối cùng: "X", "O", hoặc None nếu chưa ai thắng.
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm terminal(board)</i><br>
&emsp;&emsp;Hàm terminal(board) được sử dụng để xác định xem trò chơi đã kết thúc hay chưa. Đây là điều kiện dừng quan trọng trong thuật toán Minimax. Trò chơi được xem là kết thúc nếu có người thắng (tức winner(board) trả về “X” hoặc “O”) hoặc nếu toàn bộ các ô trên bàn cờ đều đã được đánh mà không còn ô trống nào, dẫn đến kết quả hòa. Khi hàm này trả về True, Minimax sẽ không mở rộng thêm các trạng thái con và sẽ chuyển sang tính điểm đánh giá thông qua hàm utility(). Việc kiểm tra trạng thái kết thúc nhanh và chính xác giúp thuật toán hoạt động hiệu quả và tránh duyệt không cần thiết.<br>
</div>
</font>

```python
def terminal(board):                                      #Hàm xác định xem trò chơi đã kết thúc hay chưa.
    """
    Returns True if game is over, False otherwise.
    """
    if winner(board) != None:                             #Nếu hàm winner() trả về "X" hoặc "O" → có người thắng, trò chơi kết thúc.
        return True                                       #Trả về True vì đã có người chiến thắng.

    for i in board:                                       #Duyệt từng hàng trong bàn cờ.
        for j in i:                                       #Duyệt từng ô trong hàng.
            if j == EMPTY:                                #Nếu tìm thấy bất kỳ ô nào còn trống → trò chơi chưa kết thúc.
                return False                              #Trả về False vì vẫn còn nước đi hợp lệ.

    return True                                           #Nếu không có ô trống và không có ai thắng → trò chơi kết thúc (hòa).
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm utility(board)</i><br>
&emsp;&emsp;Hàm utility(board) là một phần quan trọng trong thuật toán Minimax, dùng để gán điểm đánh giá cho các trạng thái kết thúc của trò chơi. Khi trò chơi đã đi đến trạng thái cuối cùng — tức có người thắng hoặc hòa — Minimax sẽ không mở rộng thêm mà thay vào đó tính điểm dựa trên hàm utility. Nếu người chơi “X” thắng, hàm trả về giá trị +1, thể hiện trạng thái có lợi cho MAX. Nếu người chơi “O” thắng, hàm trả về -1, thể hiện lợi thế cho MIN. Trong trường hợp hòa, hàm trả về 0. Các giá trị này sẽ được lan ngược lên trong cây quyết định của Minimax để lựa chọn nước đi tối ưu. Nhờ utility, thuật toán có cơ sở định lượng để đánh giá trạng thái cuối cùng và đưa ra chiến lược tối ưu cho AI.<br>
</div>
</font>

```python
def utility(board):                                       #Hàm trả về giá trị đánh giá cho trạng thái kết thúc của trò chơi.
    """
    Returns 1 if X has won the game, -1 if O has won, 0 otherwise.
    """
    winner_val = winner(board)                            #Gọi hàm winner() để xác định ai thắng (X, O) hoặc None nếu hòa.

    if winner_val == X:                                   #Nếu người thắng là "X" (người chơi MAX).
        return 1                                          #Trả về +1 thể hiện trạng thái có lợi cho MAX.

    elif winner_val == O:                                 #Nếu người thắng là "O" (người chơi MIN).
        return -1                                         #Trả về -1 thể hiện trạng thái có lợi cho MIN.

    return 0                                              #Nếu không ai thắng (hòa), trả về 0.
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm maxValue(state) và minValue(state)</i><br>
&emsp;&emsp;Hai hàm maxValue(state) và minValue(state) là lõi của thuật toán Minimax, chịu trách nhiệm mô phỏng quá trình suy nghĩ của hai người chơi đối kháng. Hàm maxValue đại diện cho người chơi MAX (thường là “X”), luôn tìm cách tối đa hóa điểm số bằng việc chọn nước đi tốt nhất trong các trạng thái con. Ngược lại, hàm minValue đại diện cho người chơi MIN (thường là “O”), với mục tiêu giảm thiểu điểm số của MAX bằng cách chọn nước đi gây bất lợi nhất cho đối thủ. Cả hai hàm đều hoạt động dưới dạng đệ quy: khi gặp trạng thái kết thúc (terminal state), chúng trả về giá trị đánh giá từ hàm utility(); còn khi ở trạng thái chưa kết thúc, chúng sẽ duyệt tất cả các hành động khả thi, sinh ra trạng thái mới bằng hàm result(), rồi tiếp tục gọi đệ quy lẫn nhau. Sự phối hợp qua lại giữa maxValue và minValue giúp Minimax đánh giá toàn bộ cây trạng thái trò chơi và đưa ra quyết định tối ưu cho AI.<br>
</div>
</font>

```python
def maxValue(state):                                     #Hàm đại diện cho người chơi MAX (thường là X), cố gắng chọn nước đi mang lại điểm cao nhất.
    if terminal(state):                                  #Nếu trạng thái hiện tại là trạng thái kết thúc.
        return utility(state)                            #Trả về điểm đánh giá từ hàm utility() (1, -1 hoặc 0).

    v = -math.inf                                        #Khởi tạo giá trị tốt nhất bằng -∞ để dễ dàng so sánh và tìm giá trị lớn hơn.
    for action in actions(state):                        #Duyệt tất cả các nước đi hợp lệ từ trạng thái hiện tại.
        v = max(v, minValue(result(state, action)))      #Gọi minValue cho trạng thái mới → MAX chọn giá trị lớn nhất trong các lựa chọn.
    return v                                             #Trả về giá trị tốt nhất tìm được cho MAX.

def minValue(state):                                     #Hàm đại diện cho người chơi MIN (thường là O), cố gắng chọn nước đi khiến MAX nhận điểm thấp nhất.
    if terminal(state):                                  #Nếu đây là trạng thái kết thúc.
        return utility(state)                            #Trả về điểm đánh giá (1, -1, 0).

    v = math.inf                                         #Khởi tạo giá trị tốt nhất bằng +∞ để dễ dàng tìm giá trị nhỏ hơn.
    for action in actions(state):                        #Duyệt toàn bộ các hành động hợp lệ.
        v = min(v, maxValue(result(state, action)))      #Gọi maxValue cho trạng thái mới → MIN chọn giá trị nhỏ nhất để hạn chế lợi thế MAX.
    return v                                             #Trả về giá trị tốt nhất tìm được cho MIN.
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm minimax(board)</i><br>
&emsp;&emsp;Hàm minimax(board) là chức năng trung tâm của toàn bộ thuật toán, chịu trách nhiệm lựa chọn nước đi tối ưu cho AI dựa trên đánh giá của maxValue và minValue. Dựa vào lượt chơi hiện tại (được xác định bởi hàm player(board)), hàm sẽ quyết định xem nên đóng vai MAX hay MIN. Nếu đến lượt người chơi MAX (thường là X), hàm sẽ duyệt toàn bộ các hành động hợp lệ và gọi minValue() để đánh giá kết quả của chúng, sau đó chọn hành động có giá trị lớn nhất. Ngược lại, nếu là lượt người chơi MIN (thường là O), hàm sẽ gọi maxValue() để chọn hành động có giá trị nhỏ nhất. Nhờ việc kết hợp hai hàm đệ quy này, minimax() tìm được chiến lược chơi tối ưu cho AI, đảm bảo rằng trong mọi tình huống AI không thể thua và luôn đưa ra nước đi hợp lý nhất.<br>
</div>
</font>

```python
def minimax(board):                                      #Hàm đưa ra nước đi tối ưu cho người chơi hiện tại dựa trên thuật toán Minimax.
    """
    Returns the optimal action for the current player on the board.
    """
    current_player = player(board)                       #Xác định người chơi hiện tại (user hoặc AI), từ đó biết nên dùng MAX hay MIN.

    if current_player == X:                              #Nếu người chơi hiện tại là X → vai trò MAX.
        min = -math.inf                                  #Khởi tạo giá trị tốt nhất bằng -∞ để tìm giá trị lớn nhất.
        for action in actions(board):                    #Duyệt từng nước đi hợp lệ trên bàn cờ.
            check = minValue(result(board, action))      #Gọi minValue cho trạng thái mới → đánh giá hành động theo quan điểm MAX.
            if check > min:                              #Nếu điểm đánh giá của nước đi này lớn hơn điểm tốt nhất hiện tại.
                min = check                              #Cập nhật giá trị tốt nhất.
                move = action                            #Lưu lại nước đi tối ưu cho MAX.

    else:                                                #Nếu người chơi hiện tại là O → vai trò MIN.
        max = math.inf                                   #Khởi tạo giá trị tốt nhất bằng +∞ để tìm giá trị nhỏ nhất.
        for action in actions(board):                    #Duyệt toàn bộ các nước đi có thể.
            check = maxValue(result(board, action))      #Gọi maxValue để xem MAX có thể đạt được bao nhiêu ở nhánh này.
            if check < max:                              #Nếu giá trị này nhỏ hơn điểm tốt nhất đang có (theo quan điểm MIN).
                max = check                              #Cập nhật giá trị tối ưu của MIN.
                move = action                            #Ghi lại nước đi tối ưu cho MIN.

    return move                                          #Trả về nước đi tối ưu tìm được.
```

Đây là nơi bắt đầu chạy chương trình, xử lý việc chọn người chơi, theo dõi lượt đi, cập nhật bàn cờ và tích hợp thuật toán Minimax để máy tính đưa ra nước đi tối ưu. Vòng lặp liên tục kiểm tra xem trò chơi đã kết thúc hay chưa thông qua hàm terminal(board). Nếu chưa kết thúc, chương trình xác định người chơi hiện tại và thực hiện hành động tương ứng: người dùng nhập vị trí muốn đánh, còn máy tính sử dụng minimax(board) để chọn nước đi tối ưu. Bàn cờ được cập nhật sau mỗi lượt và in ra cho người chơi quan sát. Khi có người thắng hoặc hết ô trống, vòng lặp dừng lại và thông báo kết quả. Phần này kết nối toàn bộ các hàm đã xây dựng trước đó, tạo thành một chương trình hoàn chỉnh có tương tác giữa người và máy.<br>

```python
if __name__ == "__main__":                                #Điểm bắt đầu của chương trình khi file được chạy trực tiếp.
    board = initial_state()                               #Khởi tạo bàn cờ rỗng 3×3 cho ván chơi mới.
    ai_turn = False                                       #Biến cờ đánh dấu xem có phải lượt máy hay không.

    print("Choose a player")                              #Yêu cầu người dùng chọn ký hiệu chơi.
    user = input()                                        #Người dùng nhập 'X' hoặc 'O'.

    if user == "X":                                       #Nếu người dùng chọn X → máy sẽ là O.
        ai = "O"
    else:                                                 #Ngược lại, nếu người dùng là O → máy là X.
        ai = "X"

    while True:                                           #Vòng lặp chính của trò chơi, tiếp tục cho đến khi có kết quả.
        game_over = terminal(board)                       #Kiểm tra trạng thái trò chơi đã kết thúc hay chưa.
        playr = player(board)                             #Xác định người chơi hiện tại (ai đang đến lượt).

        if game_over:                                     #Nếu trò chơi kết thúc.
            winner = winner(board)                        #Xác định ai thắng.
            if winner is None:                            #Nếu winner = None → hòa.
                print("Game Over: Tie.")
            else:
                print(f"Game Over: {winner} wins.")       #In ra người chiến thắng.
            break                                          #Thoát vòng lặp vì trận đấu đã kết thúc.

        else:                                             #Nếu trò chơi chưa kết thúc.
            if user != playr and not game_over:           #Nếu không phải lượt người dùng → là lượt của AI.
                if ai_turn:                               #Nếu biến cờ xác nhận đến lượt máy.
                    move = minimax(board)                 #Máy tính tính toán nước đi tối ưu bằng Minimax.
                    board = result(board, move)           #Cập nhật bàn cờ sau khi máy đánh.
                    ai_turn = False                       #Chuyển lượt lại cho người chơi.
                print(numpy.array(board))                 #In bàn cờ cho người dùng xem.

            elif user == playr and not game_over:         #Nếu là lượt người chơi.
                ai_turn = True                            #Chuẩn bị cho lượt tiếp theo của AI.
                print("Enter the position to move (row,col)")   #Yêu cầu người dùng nhập vị trí đánh.
                i = int(input("Row:"))                    #Nhập hàng.
                j = int(input("Col:"))                    #Nhập cột.

                if board[i][j] == EMPTY:                  #Chỉ cho phép đánh vào ô trống.
                    board = result(board, (i, j))         #Cập nhật bàn cờ sau khi người dùng đánh.
                print(numpy.array(board))                 #In bàn cờ sau lượt người chơi.
```

<b>3.1. Phân tích thuật toán Alpha-Beta</b><br>
&emsp;&emsp;<i>Hàm GetWinner(board)</i><br>
&emsp;&emsp;Hàm GetWinner(board) có nhiệm vụ kiểm tra xem trên bàn cờ hiện tại có người chơi nào giành chiến thắng hay chưa. Vì bàn cờ Tic-Tac-Toe được biểu diễn dưới dạng danh sách một chiều gồm 9 phần tử (tương ứng với 9 ô từ 1 đến 9), hàm này kiểm tra lần lượt ba điều kiện: chiến thắng theo hàng ngang, chiến thắng theo cột dọc, và chiến thắng theo đường chéo. Nếu bất kỳ tổ hợp nào trong số đó có cùng một giá trị (“X” hoặc “O”) và không phải là số (tức đã được đánh), người thắng sẽ được trả về. Nếu không có tổ hợp nào trùng khớp, hàm trả về None, nghĩa là chưa có người thắng. Đây là khối logic rất quan trọng để thuật toán Alpha-Beta biết khi nào trạng thái đang xét là trạng thái kết thúc và trả điểm đánh giá thích hợp.<br>
</div>

```python
def GetWinner(board):                                         #Hàm nhận vào một danh sách 9 phần tử đại diện cho bàn cờ 3x3.
    """
    Returns the winner in the current board if there is one, otherwise it
    returns None.
    """

    # horizontal                                                #Kiểm tra 3 hàng ngang xem có người thắng không.
    if board[0] == board[1] and board[1] == board[2]:           #Hàng 1: nếu 3 ô (0,1,2) giống nhau → có người thắng.
        return board[0]
    elif board[3] == board[4] and board[4] == board[5]:         #Hàng 2: kiểm tra các ô (3,4,5).
        return board[3]
    elif board[6] == board[7] and board[7] == board[8]:         #Hàng 3: kiểm tra các ô (6,7,8).
        return board[6]

    # vertical                                                   #Kiểm tra 3 cột dọc xem có người thắng.
    elif board[0] == board[3] and board[3] == board[6]:         #Cột 1: các ô (0,3,6) giống nhau?
        return board[0]
    elif board[1] == board[4] and board[4] == board[7]:         #Cột 2: các ô (1,4,7).
        return board[1]
    elif board[2] == board[5] and board[5] == board[8]:         #Cột 3: các ô (2,5,8).
        return board[2]

    # diagonal                                                   #Kiểm tra 2 đường chéo.
    elif board[0] == board[4] and board[4] == board[8]:         #Đường chéo chính: (0,4,8).
        return board[0]
    elif board[2] == board[4] and board[4] == board[6]:         #Đường chéo phụ: (2,4,6).
        return board[2]
```

&emsp;&emsp;<i>Hàm PrintBoard(board)</i><br>

Vì chương trình lưu trạng thái bàn cờ dưới dạng một danh sách gồm 9 phần tử (1–9 cho ô trống hoặc ký hiệu “X”, “O” cho ô đã được đánh), hàm này sắp xếp lại các phần tử thành dạng bảng 3×3 để người chơi dễ quan sát. Trước khi in bàn cờ, hàm tiến hành xóa màn hình console để đảm bảo mỗi lần hiển thị là một giao diện sạch, giúp người chơi không bị nhầm lẫn. Việc in bàn cờ sau mỗi lượt đi là bước quan trọng trong quá trình tương tác, giúp người chơi theo dõi diễn biến trận đấu, kiểm tra nước đi của AI, cũng như xác định khi nào trò chơi kết thúc.<br>
</div>

<div align="justify">
&emsp;&emsp;<i>Hàm GetAvailableCells(board)</i><br>
&emsp;&emsp;Hàm GetAvailableCells(board) có nhiệm vụ xác định tất cả các ô còn trống trên bàn cờ để sử dụng cho việc sinh nước đi mới trong thuật toán Minimax kết hợp Alpha-Beta Pruning. Vì bàn cờ được biểu diễn dưới dạng một danh sách gồm 9 phần tử, trong đó các ô trống mang giá trị số (1–9) và các ô đã đánh mang ký hiệu "X" hoặc "O", hàm này kiểm tra từng phần tử của danh sách và thu thập lại những ô chưa bị chiếm. Danh sách các ô trống này được dùng trong hai trường hợp: để người chơi nhập nước đi hợp lệ và để thuật toán AI duyệt các trạng thái con trong cây tìm kiếm. Việc xác định chính xác các ô trống giúp thuật toán hoạt động đúng và tránh lỗi truy cập các ô đã được đánh.<br>
</div>
</font>

```python
def GetAvailableCells(board):                               #Hàm trả về danh sách các ô còn trống trong bàn cờ.
    """
    Returns a list of indices containing all available cells in a board.
    """

    available = list()                                      #Khởi tạo danh sách rỗng để lưu các ô có thể đánh.

    for cell in board:                                      #Duyệt qua từng phần tử trong danh sách bàn cờ.
        if cell != "X" and cell != "O":                     #Nếu giá trị ô KHÔNG phải là X hoặc O → nghĩa là ô trống.
            available.append(cell)                          #Thêm ô đó vào danh sách các ô trống.

    return available                                        #Trả về danh sách toàn bộ ô còn trống (là các số 1–9).
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm minimax(position, depth, alpha, beta, isMaximizing)</i><br>
&emsp;&emsp;Hàm minimax(position, depth, alpha, beta, isMaximizing) là trung tâm của thuật toán Minimax kết hợp tỉa nhánh Alpha-Beta, giúp AI đánh giá mọi trạng thái có thể xảy ra trong trò chơi để chọn nước đi tối ưu nhất. Thuật toán hoạt động bằng cách mô phỏng tất cả nước đi có thể, nhưng nhờ hai giá trị alpha và beta, nhiều nhánh không cần thiết sẽ được cắt bỏ nhằm tăng tốc xử lý. Nếu đang ở lượt MAX (thường là “X”), thuật toán sẽ cố gắng chọn giá trị lớn nhất trong các trạng thái con, trong khi MIN (“O”) sẽ chọn giá trị nhỏ nhất nhằm hạn chế lợi thế của MAX. Hàm còn sử dụng tham số depth để ưu tiên chiến thắng nhanh hơn khi thắng hoặc kéo dài trận đấu khi thua, giúp AI chơi thông minh và thực tế hơn. Kết quả trả về là một giá trị đánh giá thể hiện chất lượng của trạng thái, cho phép hàm FindBestMove tìm được nước đi mạnh nhất.<br>
</div>
</font>

```python
def minimax(position, depth, alpha, beta, isMaximizing):          #Hàm minimax sử dụng alpha-beta pruning để đánh giá trạng thái và chọn nước đi tối ưu.
    """
    The AI algorithm responsible for choosing the best move. Returns best
    value of a move.
    """

    winner = GetWinner(position)                                   #Kiểm tra xem trạng thái hiện tại có người thắng chưa.
    if winner != None:                                             #Nếu có người thắng → đây là trạng thái kết thúc.
        return 10 - depth if winner == "X" else -10 + depth        #Trả về điểm đánh giá: thắng nhanh thì điểm cao, thua chậm thì điểm ít bị phạt.

    if len(GetAvailableCells(position)) == 0:                      #Nếu không còn ô trống → hòa.
        return 0                                                   #Trạng thái hòa có điểm 0.

    # ========================= MAXIMIZING PLAYER (X) =========================
    if isMaximizing:                                               #Nếu hiện tại là lượt của MAX (X).
        maxEval = -math.inf                                        #Khởi tạo giá trị tốt nhất bằng -∞ để tìm giá trị lớn nhất.

        for cell in GetAvailableCells(position):                   #Duyệt từng ô còn trống để thử nước đi.
            position[cell - 1] = "X"                               #ĐẶT tạm quân X vào ô đang xét.
            Eval = minimax(position, depth + 1, alpha, beta, False) #Gọi đệ quy để đánh giá sau khi X đánh → chuyển lượt cho MIN.
            maxEval = max(maxEval, Eval)                           #Cập nhật giá trị tốt nhất mà MAX có thể đạt được.
            alpha = max(alpha, Eval)                               #Cập nhật alpha (giá trị đảm bảo của MAX).
            position[cell - 1] = cell                              #HOÀN TÁC nước đi → trả ô về giá trị ban đầu.

            if beta <= alpha:                                      #Điều kiện tỉa nhánh: nếu MIN sẽ tránh nhánh này.
                break                                              #Không cần duyệt tiếp → cắt tỉa để tăng tốc.

        return maxEval                                             #Trả về điểm tốt nhất cho MAX.

    # ========================= MINIMIZING PLAYER (O) =========================
    else:                                                          #Nếu hiện tại là lượt của MIN (O).
        minEval = +math.inf                                        #Khởi tạo giá trị tồi nhất bằng +∞ để tìm giá trị nhỏ nhất.

        for cell in GetAvailableCells(position):                   #Duyệt tất cả các ô trống.
            position[cell - 1] = "O"                               #ĐẶT tạm quân O vào ô đang xét.
            Eval = minimax(position, depth + 1, alpha, beta, True) #Chuyển lượt lại cho MAX.
            minEval = min(minEval, Eval)                           #MIN chọn giá trị thấp nhất có thể để giảm lợi thế MAX.
            beta = min(beta, Eval)                                 #Cập nhật beta (giá trị đảm bảo của MIN).
            position[cell - 1] = cell                              #HOÀN TÁC nước đi.

            if beta <= alpha:                                      #Nếu alpha đã vượt beta → cắt tỉa nhánh.
                break

        return minEval                                             #Trả về điểm tốt nhất cho MIN.
```

<font face="Times New Roman">
<div align="justify">
&emsp;&emsp;<i>Hàm FindBestMove(currentPosition, AI)</i><br>
&emsp;&emsp;Hàm FindBestMove(currentPosition, AI) chịu trách nhiệm tìm nước đi tối ưu cho máy (AI) dựa trên kết quả đánh giá của hàm minimax có alpha-beta pruning. Hàm sẽ duyệt qua tất cả các ô trống trên bàn cờ, thử đặt quân của AI vào từng ô, sau đó gọi hàm minimax() để tính giá trị của nước đi đó. Nếu AI đóng vai trò MAX (thường là "X"), nó sẽ chọn nước đi có giá trị minimax lớn nhất; còn nếu là MIN ("O"), nó chọn nước đi có giá trị nhỏ nhất. Sau khi đánh giá tất cả các ô hợp lệ, hàm trả về ô mang lại lợi thế tối ưu nhất cho AI. Đây là hàm then chốt kết nối giữa thuật toán minimax và hành vi thực tế của AI trong game, quyết định việc AI chơi thông minh, không mắc sai lầm và không thể bị đánh bại.<br>
</div>
</font>

```python
def FindBestMove(currentPosition, AI):                         #Hàm tìm nước đi tốt nhất cho AI dựa trên đánh giá minimax.
    """
    This will return the best possible move for the player.
    Will Traverse all cells, evaluate minimax function for all empty cells.
    And return the cell with optimal value.
    """

    bestVal = -math.inf if AI == "X" else +math.inf            #Nếu AI là X (MAX) → tìm giá trị lớn nhất; nếu AI là O (MIN) → tìm nhỏ nhất.
    bestMove = -1                                              #Biến lưu ô tốt nhất, mặc định là -1 (chưa có).

    for cell in GetAvailableCells(currentPosition):            #Duyệt từng ô trống trong bàn cờ.
        currentPosition[cell - 1] = AI                         #THỬ đặt quân AI vào ô đó.

        moveVal = minimax(                                     #Gọi minimax để đánh giá điểm của nước đi này.
            currentPosition,                                   #Trạng thái bàn cờ sau khi thử nước đi.
            0,                                                 #Độ sâu bắt đầu là 0 vì đây là lượt gốc.
            -math.inf,                                         #Alpha khởi tạo bằng -∞.
            +math.inf,                                         #Beta khởi tạo bằng +∞.
            False if AI == "X" else True                       #Nếu AI là X → lượt MIN tiếp theo; nếu AI là O → lượt MAX.
        )

        currentPosition[cell - 1] = cell                       #HOÀN TÁC nước đi, trả ô về giá trị ban đầu (1–9).

        if (AI == "X" and moveVal > bestVal):                  #Nếu AI là MAX (X) → tìm điểm lớn nhất.
            bestMove = cell                                    #Cập nhật ô tốt nhất.
            bestVal = moveVal                                  #Cập nhật giá trị tốt nhất.

        elif (AI == "O" and moveVal < bestVal):                #Nếu AI là MIN (O) → tìm điểm nhỏ nhất.
            bestMove = cell                                    #Cập nhật ô tốt nhất.
            bestVal = moveVal                                  #Cập nhật giá trị tốt nhất.

    return bestMove                                             #Trả về ô mang giá trị tốt nhất cho AI.
```

&emsp;&emsp;<i>Hàm main()</i><br>
&emsp;&emsp;Hàm main() giữ vai trò điều khiển toàn bộ logic của trò chơi, bao gồm việc nhận lựa chọn của người chơi, xác định lượt đi, cập nhật bàn cờ, gọi thuật toán Alpha-Beta để AI đưa ra nước đi tối ưu và kiểm tra trạng thái thắng–thua sau mỗi lượt. Trò chơi bắt đầu bằng việc người chơi chọn “X” hoặc “O”, sau đó AI được gán ký hiệu còn lại. Vì “X” luôn đi trước, vòng lặp chính liên tục luân chuyển lượt giữa người và máy cho đến khi có người thắng hoặc bàn cờ đầy. Nếu đến lượt AI, chương trình sẽ gọi FindBestMove() để tìm nước đi mạnh nhất theo thuật toán minimax với alpha-beta pruning. Nếu đến lượt người chơi, chương trình yêu cầu nhập ô hợp lệ rồi cập nhật bàn cờ. Sau mỗi lượt, hàm kiểm tra xem có người thắng hay không và kết thúc trò chơi khi điều kiện xảy ra. Nhờ vòng điều khiển chính này, tất cả các thành phần trong chương trình được liên kết lại để tạo nên một trò chơi Tic-Tac-Toe hoàn chỉnh có AI bất bại.<br>
</div>
</font>

```python
def main():                                                    #Hàm chính điều khiển toàn bộ trò chơi Tic-Tac-Toe.
    player = input("Play as X or O? ").strip().upper()         #Yêu cầu người chơi nhập ký hiệu muốn sử dụng (X hoặc O).
    AI = "O" if player == "X" else "X"                         #AI sẽ nhận ký hiệu còn lại: nếu người chơi là X → AI là O, ngược lại.

    currentGame = [*range(1, 10)]                              #Khởi tạo bàn cờ: danh sách 1–9 biểu thị các ô trống ban đầu.

    # X always starts first.                                   #Quy tắc: X luôn là người đi trước.
    currentTurn = "X"                                          #Biến đánh dấu lượt chơi hiện tại.
    counter = 0                                                #Bộ đếm số lượt đi (tối đa 9 nước).

    while True:                                                #Vòng lặp chính của trò chơi.
        if currentTurn == AI:                                  #Nếu đến lượt AI đi.
            # NOTE: if the AI starts first, it'll always choose index 0 so
            # to save time you could play it.                  #Ghi chú: Nếu AI là X và bắt đầu trước, có chiến lược cố định.

            cell = FindBestMove(currentGame, AI)               #AI gọi thuật toán minimax alpha-beta để tìm nước đi tối ưu.
            currentGame[cell - 1] = AI                         #Cập nhật bàn cờ với nước đi của AI.
            currentTurn = player                               #Chuyển lượt cho người chơi.

        elif currentTurn == player:                            #Nếu đến lượt người chơi.
            PrintBoard(currentGame)                            #In bàn cờ để người chơi quan sát.
            while True:                                        #Vòng lặp để đảm bảo người chơi nhập ô hợp lệ.
                humanInput = int(input("Enter Number: ").strip())  #Nhập ô muốn đánh.
                if humanInput in currentGame:                  #Nếu ô đó còn trống (vẫn là số).
                    currentGame[humanInput - 1] = player       #Đặt ký hiệu người chơi vào ô đó.
                    currentTurn = AI                           #Chuyển lượt lại cho AI.
                    break                                      #Thoát vòng lặp nhập.
                else:
                    PrintBoard(currentGame)                    #Nếu nhập sai → in lại bàn cờ.
                    print("Cell Not Available.")               #Thông báo lỗi.

        if GetWinner(currentGame) != None:                     #Sau mỗi lượt → kiểm tra xem có người thắng chưa.
            PrintBoard(currentGame)                            #In kết quả bàn cờ cuối cùng.
            print(f"{GetWinner(currentGame)} WON!!!")          #Thông báo người thắng.
            break                                              #Kết thúc trò chơi.

        counter += 1                                           #Tăng số lượt đi đã thực hiện.

        if GetWinner(currentGame) == None and counter == 9:    #Nếu không ai thắng và đủ 9 lượt → hòa.
            PrintBoard(currentGame)
            print("Tie.")
            break                                              #Kết thúc trò chơi.
```
