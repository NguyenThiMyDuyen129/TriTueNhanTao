<b>CHƯƠNG 1: CƠ SỞ LÝ THUYẾT RÀNG BUỘC TRONG THUẬT TOÁN</b>
</center>

<p>
<div align="justify">
<b>1.1. Giới thiệu ràng buộc</b><br>
&emsp;&emsp;Ràng buộc trong thuật toán được hiểu là một quan hệ áp đặt lên một tập các biến, nhằm quy định những giá trị nào của các biến là hợp lệ. Các ràng buộc này giúp thu hẹp không gian nghiệm và đảm bảo lời giải thỏa mãn các điều kiện đã đặt ra. Việc biểu diễn ràng buộc có thể thực hiện theo nhiều cách khác nhau, phổ biến nhất là dùng các biểu thức toán học hoặc logic để mô tả quan hệ giữa các biến, hoặc sử dụng bảng liệt kê tất cả các tổ hợp giá trị hợp lệ mà các biến có thể nhận. Nhờ có ràng buộc, thuật toán có thể kiểm tra, loại bỏ hoặc đánh giá các nghiệm tiềm năng một cách hiệu quả, từ đó tìm ra lời giải đúng và phù hợp với yêu cầu của bài toán.<br>
<b>1.2. Bài toán thỏa mãn ràng buộc</b><br>
&emsp;&emsp;Bài toán thỏa mãn ràng buộc (Constraint Satisfaction Problem – CSP) là loại bài toán trong đó mục tiêu là tìm một bộ giá trị gán cho các biến sao cho tất cả các ràng buộc được đáp ứng đồng thời. Một CSP gồm ba thành phần chính: tập biến, miền giá trị của mỗi biến, và tập ràng buộc chi phối các biến đó. Nhiệm vụ của thuật toán là xác định nghiệm thỏa mãn toàn bộ các điều kiện, hoặc kết luận rằng không tồn tại nghiệm hợp lệ. Các bài toán CSP thường xuất hiện trong lập lịch, phân công công việc, tô màu đồ thị, sudoku, tối ưu hóa tài nguyên và nhiều bài toán trí tuệ nhân tạo.<br>
<b>1.2.1. Phương pháp giải quyết bài toán thỏa mãn ràng buộc</b><br>
&emsp;&emsp;CSP có thể được giải quyết bằng nhiều kỹ thuật khác nhau, bao gồm:<br>
&emsp;&emsp;- Quay lui: Một thuật toán tìm kiếm theo chiều sâu mà xây dựng dần các ứng viên cho các giải pháp và từ bỏ một ứng viên ngay khi xác định rằng nó không thể dẫn đến một giải pháp hợp lệ.<br>
&emsp;&emsp;- Lan truyền ràng buộc: Kỹ thuật này giảm không gian tìm kiếm bằng cách áp dụng các ràng buộc cục bộ, điều này có thể đơn giản hóa vấn đề trước khi áp dụng các thuật toán tìm kiếm.<br>
&emsp;&emsp;- Phương pháp Heuristic: Những phương pháp này sử dụng các quy tắc kinh nghiệm để hướng dẫn quá trình tìm kiếm, cải thiện hiệu quả.<br>
&emsp;&emsp;Các tiến bộ gần đây trong CSPs tận dụng các kỹ thuật học máy để nâng cao khả năng giải quyết vấn đề, cho phép xử lý các ràng buộc phức tạp một cách hiệu quả hơn. Ví dụ, việc tích hợp CSPs với mạng nơ-ron có thể dẫn đến các giải pháp sáng tạo trong các môi trường động.<br>
</p>
<b>1.2.2. Vai trò của bài toán thỏa mãn ràng buộc</b><br>
&emsp;&emsp;CSP được sử dụng rộng rãi trong các ứng dụng thực tế như:<br>
&emsp;&emsp;- Lập lịch: Phân bổ tài nguyên cho các nhiệm vụ theo thời gian trong khi tôn trọng các ràng buộc như thời hạn và khả năng sẵn có của tài nguyên.<br>
&emsp;&emsp;- Phân bổ tài nguyên: Phân phối tài nguyên hạn chế giữa các nhiệm vụ hoặc tác nhân cạnh tranh một cách hiệu quả.<br>
&emsp;&emsp;- Lập luận tự động: Cho phép hệ thống suy luận thông tin mới từ các sự kiện và ràng buộc đã biết.<br>

<b>CHƯƠNG 3: PHÂN TÍCH, TRIỂN KHAI VÀ CÀI ĐẶT THUẬT TOÁN TÔ MÀU ĐỒ THỊ</b>

<div align="justify">
<b>3.1. Cơ sở lý thuyết thuật toán tô màu</b><br>
&emsp;&emsp;Đồ thị (Graph) là một mô hình toán học dùng để biểu diễn mối quan hệ giữa các đối tượng. Một đồ thị được định nghĩa là một tập các đối tượng được gọi là các đỉnh (hoặc nút) nối với nhau bởi các cạnh (hoặc cung). Cạnh có thể có hướng hoặc vô hướng. Đồ thị thường được vẽ dưới dạng một tập các điểm (các đỉnh nối với nhau bằng các đoạn thẳng (các cạnh)).<br>
&emsp;&emsp;Đồ thị biểu diễn được rất nhiều cấu trúc, nhiều bài toán thực tế có thể được biểu diễn bằng đồ thị. Ví dụ, trong mạng xã hội, mỗi người dùng là một đỉnh, các cạnh thể hiện mối quan hệ bạn bè giữa hai người. Trong hệ thống giao thông, mỗi thành phố là một đỉnh, còn các con đường nối giữa chúng là các cạnh.<br>
&emsp;&emsp;Đồ thị có trọng số (Weighted Graph): Cấu trúc đồ thị có thể được mở rộng bằng cách gán trọng số cho mỗi cạnh. Có thể sử dụng đồ thị có trọng số để biểu diễn các khái niệm khác nhau. Ví dụ, trọng số thể hiện một đại lượng định lượng như khoảng cách, thời gian, chi phí hoặc năng lượng. Trong bản đồ giao thông, trọng số có thể là khoảng cách giữa hai thành phố hoặc thời gian di chuyển giữa hai vị trí.<br>
&emsp;&emsp;Đồ thị có hướng (Directed Graph): Một cách khác để mở rộng đồ thị cơ bản là quy định hướng cho các cạnh của đồ. Khi đó, mỗi cạnh mang thông tin về chiều đi từ một đỉnh này đến đỉnh khác (Ví dụ, trong mạng thông tin hoặc giao thông một chiều, dữ liệu (hoặc xe cộ) chỉ có thể di chuyển theo chiều đã được quy định trước), loại đồ thị này được gọi là đồ thị có hướng. Một đồ thị có hướng với các cạnh mang trọng số được gọi là một mạng có trọng số (Weighted Directed Network).<br>
&emsp;&emsp;Đồ thị có hướng và có trọng số (Weighted Directed Graph): Đây là loại đồ thị kết hợp cả hai yếu tố hướng và trọng số. Mỗi cạnh không chỉ có hướng di chuyển mà còn mang một giá trị cụ thể thể hiện chi phí hoặc độ ưu tiên. Ví dụ: Trong mạng máy tính, các đỉnh biểu diễn thiết bị (router, server), còn các cạnh thể hiện đường truyền dữ liệu.<br>
&emsp;&emsp;Trọng số có thể là tốc độ truyền hoặc độ trễ mạng. Khi đó, việc tìm đường truyền tối ưu giữa hai máy chính là bài toán đường đi ngắn nhất trên đồ thị có hướng và có trọng số.<br>
&emsp;&emsp;Trong toán học rời rạc và khoa học máy tính, vấn đề tô màu đồ thị (Graph Coloring) là một trong những bài toán cơ bản nhưng có tầm quan trọng đặc biệt trong lý thuyết đồ thị. Mục tiêu của bài toán là gán màu cho các đỉnh của đồ thị sao cho không có hai đỉnh kề nhau nào có cùng màu, đồng thời sử dụng số màu ít nhất có thể. Đây là một dạng bài toán tối ưu hóa tổ hợp, được ứng dụng rộng rãi trong các lĩnh vực như lập lịch công việc, phân bổ tài nguyên, tối ưu hóa mạng, và thiết kế hệ thống truyền thông không dây. <br>
&emsp;&emsp;Bài toán tô màu đồ thị bao gồm nhiều loại: tô màu đỉnh đồ thị (vertex graph coloring), tô màu cạnh đồ thị (edge graph coloring),...Thông thường, sẽ xét dạng phổ biến nhất là tô màu các đỉnh của đồ thị (Vertex Coloring).<br>
&emsp;&emsp;Quá trình tô màu đồ thị được biểu diễn bằng một ánh xạ toán học. Trong đó, G = (V,E) là đồ thị gồm tập đỉnh V(G) và tập cạnh E(G). V(G) là tập các đỉnh của đồ thị. Mỗi phần tử u, v ∈ V(G) là một nút trong mạng lưới. E(G) là tập các cạnh nối giữa các đỉnh. Mỗi cạnh được ký hiệu là (u,v), biểu thị mối liên hệ giữa đỉnh u và v.<br>
&emsp;&emsp;&emsp;&emsp;f: V(G)→{c_1, c_2, … , c_k}<br>
&emsp;&emsp;Ánh xạ f là một hàm gán màu cho từng đỉnh của đồ thị. Tập c_1, c_2, … , c_k đại diện cho k màu khác nhau có thể sử dụng trong phép tô màu. Trong đó c_i là các màu sao cho với mọi cặp đỉnh kề (u,v) ∈ E(G), f(u) ≠ f(v). Nghĩa là hai đỉnh u và v nối trực tiếp bằng một cạnh không được tô cùng màu. Mục tiêu của bài toán là tối thiểu hóa số lượng màu k được dùng trong phép tô màu, tức là tìm ra giá trị nhỏ nhất của k sao cho ánh xạ f vẫn thỏa mãn điều kiện trên. Số lượng màu nhỏ nhất cần để tô một đồ thị G theo quy tắc trên được gọi là số sắc của đồ thị, ký hiệu là χ(G) (chromatic number).<br>
&emsp;&emsp;χ(G)=min{ k ∣ G có thể được tô bằng k màu hợp lệ}<br>
&emsp;&emsp;χ(G): là số sắc (chromatic number), số màu ít nhất cần có để tô đồ thị G một cách hợp lệ.<br>
&emsp;&emsp;Đây là đại lượng đo độ phức tạp tô màu của đồ thị: đồ thị càng ràng buộc nhiều (nhiều cạnh nối giữa các đỉnh kề nhau) thì χ(G) càng lớn. Có thể hình dung qua ví dụ minh họa G = (V,E),V = {A,B,C}, E = {(A,B),(B,C)}, chọn ánh xạ màu như sau:<br>
&emsp;&emsp;&emsp;&emsp;f(A) = c_1, f(B) = c_2, f(C) = c_1. Trong đó:<br>
&emsp;&emsp;- Hai đỉnh kề nhau A–B và B–C không cùng màu.<br>
&emsp;&emsp;- Tổng cộng chỉ cần hai màu để tô toàn bộ đồ thị.<br>
<b>3.2. Tính chất và thành phần thuật toán</b><br>
&emsp;&emsp;Thuật toán tô màu tối ưu đồ thị có những tính chất đặc trưng phản ánh bản chất của quá trình gán màu cho các đỉnh trong đồ thị nhằm đạt hiệu quả tối ưu. Các tính chất chính bao gồm:<br>
&emsp;&emsp;Không tô cùng màu cho các đỉnh kề nhau: Hai đỉnh được gọi là kề nhau nếu tồn tại một cạnh nối trực tiếp giữa chúng. Thuật toán đảm bảo rằng không có hai đỉnh liền kề nào được tô cùng một màu, qua đó duy trì tính hợp lệ của phép tô màu.<br>
&emsp;&emsp;Tối thiểu hóa số lượng màu sử dụng: Mục tiêu thuật toán là sử dụng số màu nhỏ nhất nhưng vẫn đảm bảo quy tắc không trùng màu giữa các đỉnh kề. Số lượng màu nhỏ nhất cần thiết để tô một đồ thị hợp lệ được gọi là số sắc (χ(G)) của đồ thị.<br>
&emsp;&emsp;Không xét hướng của cạnh (trong đa số trường hợp): Tô màu thường áp dụng với đồ thị vô hướng, do hướng của cạnh không ảnh hưởng đến ý nghĩa liền kề. Tuy nhiên, trong một số bài toán, khái niệm có thể được mở rộng cho đồ thị có hướng.<br>
&emsp;&emsp;Tính rời rạc và tổ hợp cao: Việc gán màu cho các đỉnh là độc lập nhưng có ràng buộc bởi cấu trúc liền kề của đồ thị. Do sự kết hợp giữa các lựa chọn màu tiềm năng, bài toán mang bản chất tối ưu hóa tổ hợp và có độ phức tạp cao (NP-khó).<br>
&emsp;&emsp;Tính xấp xỉ: Trong trường hợp đồ thị lớn hoặc phức tạp, thuật toán chính xác khó khả thi. Khi đó, thuật toán heuristic hoặc metaheuristic được sử dụng nhằm tìm nghiệm gần tối ưu, đảm bảo tính hợp lệ nhưng giảm thiểu thời gian tính toán.<br>
&emsp;&emsp;Một thuật toán tô màu tối ưu đồ thị điển hình luôn bao gồm các thành phần chính sau:<br>
&emsp;&emsp;Tập đỉnh (Vertex Set – V): Tập hợp các đỉnh cần được tô màu. Mỗi đỉnh đại diện cho một đối tượng hoặc thực thể trong bài toán thực tế (ví dụ: môn thi, công việc, trạm phát sóng, tiến trình trong hệ thống,...).<br>
&emsp;&emsp;Tập cạnh (Edge Set – E): Tập hợp các cạnh thể hiện mối quan hệ giữa các đỉnh. Nếu giữa hai đỉnh có cạnh nối, điều đó có nghĩa là chúng không được phép mang cùng một màu.<br>
&emsp;&emsp;Tập màu (Color Set – C): Tập hợp các màu hoặc nhãn có thể gán cho các đỉnh. Mục tiêu của thuật toán là giảm số lượng màu sử dụng trong tập này đến mức tối thiểu mà vẫn duy trì tính hợp lệ.<br>
&emsp;&emsp;Hàm chọn màu (Color Selection Rule): Quy tắc xác định màu sẽ được gán cho mỗi đỉnh tại thời điểm tô. Quy tắc phổ biến thường được dùng, thuật toán tô màu tham lam (Greedy Coloring), chọn màu nhỏ nhất chưa sử dụng bởi các đỉnh kề. <br>
&emsp;&emsp;Điều kiện hợp lệ (Feasibility Condition): Sau mỗi lần gán màu, thuật toán phải kiểm tra xem có vi phạm quy tắc tô màu (hai đỉnh liền kề cùng màu) hay không. Nếu có xung đột, cần lựa chọn lại màu khác.<br>
&emsp;&emsp;Điều kiện dừng (Termination Condition): Thuật toán dừng lại khi tất cả các đỉnh đã được tô hợp lệ, nghĩa là không còn đỉnh nào chưa được gán màu.<br>
&emsp;&emsp;Hàm tối ưu hóa (Optimization Objective): Sau khi đã có một phép tô hợp lệ, thuật toán tiếp tục cải tiến lời giải bằng cách thử hoán đổi thứ tự tô hoặc gộp các nhóm màu tương thích để giảm thêm số màu cần thiết, hướng đến nghiệm tối ưu hoặc gần tối ưu nhất.<br>

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# ===== HÀM ĐỌC MA TRẬN KỀ TỪ FILE TXT =====
def read_adj_matrix(path):
    with open(path, 'r') as f:
        lines = f.readlines()

    matrix = []
    for line in lines:
        if line.strip() == "":
            continue
        row = [int(x) for x in line.split()]
        matrix.append(row)

    A = np.array(matrix, dtype=int)
    if A.shape[0] != A.shape[1]:
        raise ValueError("Ma trận không vuông!")
    return A

# ===== THUẬT TOÁN TÔ MÀU VÀ IN CÁC BƯỚC =====
def greedy_coloring(adj_matrix, print_steps=True):
    n = adj_matrix.shape[0]
    alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    nodes = list(alpha[:n])

    degree = [int(np.sum(adj_matrix[i])) for i in range(n)]

    if print_steps:
        print("Bậc từng đỉnh:")
        for i, d in enumerate(degree):
            print(f"  {nodes[i]}: {d}")
        print("-"*40)

    idx_sorted = sorted(range(n), key=lambda i: degree[i], reverse=True)
    sorted_nodes = [nodes[i] for i in idx_sorted]

    if print_steps:
        print("Thứ tự duyệt:", " → ".join(sorted_nodes))
        print("-"*40)

    all_colors = ["blue", "red", "yellow", "green", "orange", "purple", "brown"]
    color_of = {}
    step = 1

    for u in sorted_nodes:
        i = nodes.index(u)

        # Các màu bị cấm do láng giềng đã dùng
        forbidden = set()
        for j in range(n):
            if adj_matrix[i, j] == 1:
                v = nodes[j]
                if v in color_of:
                    forbidden.add(color_of[v])

        # Chọn màu đầu tiên không bị cấm
        chosen = next(c for c in all_colors if c not in forbidden)
        color_of[u] = chosen

        if print_steps:
            print(f"Bước {step}: xét đỉnh {u}")
            print("  Đỉnh kề đã tô:")
            for j in range(n):
                if adj_matrix[i, j] == 1:
                    v = nodes[j]
                    if v in color_of and v != u:
                        print(f"    {v}: {color_of[v]}")
            print(f"  Màu cấm: {sorted(forbidden)}")
            print(f"  → Gán màu cho {u}: {chosen}")
            print("-"*40)
        step += 1

    return nodes, color_of, degree

# ===== VẼ ĐỒ THỊ ĐÃ TÔ MÀU =====
def draw_colored_graph(adj_matrix, nodes, color_of):
    G = nx.Graph()
    G.add_nodes_from(nodes)

    n = adj_matrix.shape[0]
    for i in range(n):
        for j in range(i+1, n):
            if adj_matrix[i, j] == 1:
                G.add_edge(nodes[i], nodes[j])

    colors = [color_of[v] for v in nodes]

    plt.figure(figsize=(3, 3))
    pos = nx.spring_layout(G, seed=42)
    nx.draw(G, pos,
            with_labels=True,
            node_color=colors,
            node_size=500,
            font_color="white",
            font_weight="bold",
            edge_color="black")
    plt.title("Đồ thị sau khi tô màu")
    plt.axis("off")
    plt.show()

# ======= CHẠY THỬ VỚI FILE graph.txt VỪA TẠO =======
A = read_adj_matrix("graph.txt")
print("Ma trận kề đọc được:")
print(A)
print("="*40)

nodes, color_of, degree = greedy_coloring(A, print_steps=True)

print("KẾT QUẢ CUỐI CÙNG:")
for i, v in enumerate(nodes):
    print(f"  Đỉnh {v} (bậc {degree[i]}) → {color_of[v]}")

draw_colored_graph(A, nodes, color_of)
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# ===== HÀM ĐỌC MA TRẬN KỀ TỪ FILE TXT =====
def read_adj_matrix(path):
    with open(path, 'r') as f:
        lines = f.readlines()

    matrix = []
    for line in lines:
        if line.strip() == "":
            continue
        row = [int(x) for x in line.split()]
        matrix.append(row)

    A = np.array(matrix, dtype=int)
    if A.shape[0] != A.shape[1]:
        raise ValueError("Ma trận không vuông!")
    return A


# ===== VẼ ĐỒ THỊ Ở MỘT BƯỚC (HIỂN THỊ TRỰC TIẾP) =====
def draw_step(adj_matrix, nodes, color_of, current_node, step):
    G = nx.Graph()
    G.add_nodes_from(nodes)
    n = adj_matrix.shape[0]
    for i in range(n):
        for j in range(i+1, n):
            if adj_matrix[i, j] == 1:
                G.add_edge(nodes[i], nodes[j])

    # màu cho từng đỉnh: chưa tô -> xám
    default_color = "lightgray"
    node_colors = []
    for v in nodes:
        if v in color_of:
            node_colors.append(color_of[v])
        else:
            node_colors.append(default_color)

    plt.figure(figsize=(2, 2))
    pos = nx.spring_layout(G, seed=42)

    nx.draw(
        G, pos,
        with_labels=True,
        node_color=node_colors,
        node_size=300,
        font_color="white",
        font_weight="bold",
        edge_color="black"
    )

    # viền đỏ cho đỉnh đang xét
    nx.draw_networkx_nodes(
        G, pos,
        nodelist=[current_node],
        node_size=350,
        node_color=node_colors[nodes.index(current_node)],
        edgecolors="red",
        linewidths=3
    )

    plt.title(f"Bước {step}: tô đỉnh {current_node}")
    plt.axis("off")
    plt.show()


# ===== THUẬT TOÁN TÔ MÀU + IN BƯỚC + VẼ ẢNH NGAY DƯỚI =====
def greedy_coloring_with_inline_images(adj_matrix):
    n = adj_matrix.shape[0]
    alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    nodes = list(alpha[:n])

    degree = [int(np.sum(adj_matrix[i])) for i in range(n)]

    print("Bậc từng đỉnh:")
    for i, d in enumerate(degree):
        print(f"  {nodes[i]}: {d}")
    print("-"*40)

    idx_sorted = sorted(range(n), key=lambda i: degree[i], reverse=True)
    sorted_nodes = [nodes[i] for i in idx_sorted]
    print("Thứ tự duyệt:", " → ".join(sorted_nodes))
    print("="*40)

    all_colors = ["blue", "red", "yellow", "green", "orange", "purple", "brown"]
    color_of = {}
    step = 1

    for u in sorted_nodes:
        i = nodes.index(u)

        # màu bị cấm từ láng giềng
        forbidden = set()
        for j in range(n):
            if adj_matrix[i, j] == 1:
                v = nodes[j]
                if v in color_of:
                    forbidden.add(color_of[v])

        chosen = next(c for c in all_colors if c not in forbidden)
        color_of[u] = chosen

        # IN BƯỚC
        print(f"Bước {step}: xét đỉnh {u}")
        print("  Đỉnh kề đã tô:")
        for j in range(n):
            if adj_matrix[i, j] == 1:
                v = nodes[j]
                if v in color_of and v != u:
                    print(f"    {v}: {color_of[v]}")
        print(f"  Màu cấm: {sorted(forbidden)}")
        print(f"  → Gán màu cho {u}: {chosen}")

        # HIỂN THỊ ẢNH NGAY DƯỚI BƯỚC NÀY
        draw_step(adj_matrix, nodes, color_of, u, step)

        print("-"*40)
        step += 1

    return nodes, color_of, degree


# ====== CHẠY VỚI FILE CỦA BẠN ======
# Nếu bạn đã có file txt trong /content, sửa "graph.txt" thành tên file của bạn.
print("="*40)
A = read_adj_matrix("graph.txt")
print("Ma trận kề đọc được:")
print(A)
print("="*40)

nodes, color_of, degree = greedy_coloring_with_inline_images(A)

print("KẾT QUẢ CUỐI CÙNG:")
for i, v in enumerate(nodes):
    print(f"  Đỉnh {v} (bậc {degree[i]}) → {color_of[v]}")
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt


# ===== RANDOM MA TRẬN KỀ CÓ RÀNG BUỘC + LƯU FILE TXT =====
def generate_constrained_random_adj_matrix(n=6, p=0.3, save_path="graph.txt"):
    """
    Sinh ma trận kề NGẪU NHIÊN với các ràng buộc:
      - Không phải đồ thị đường thẳng (path graph)
      - Mỗi đỉnh có bậc >= 2
    Sau đó LƯU VÀO FILE graph.txt
    """
    while True:
        A = np.zeros((n, n), dtype=int)

        # random cạnh
        for i in range(n):
            for j in range(i+1, n):
                if np.random.rand() < p:
                    A[i, j] = 1
                    A[j, i] = 1

        # tính bậc
        degree = [int(np.sum(A[i])) for i in range(n)]

        # Ràng buộc 1: tất cả bậc ≥ 2
        if any(d < 2 for d in degree):
            continue

        # Ràng buộc 2: không phải đồ thị đường thẳng
        if degree.count(1) == 2 and degree.count(2) == n - 2:
            continue

        # Nếu hợp lệ → LƯU FILE
        np.savetxt(save_path, A, fmt="%d", delimiter=" ")

        print(f"✔ Đã lưu ma trận hợp lệ vào file: {save_path}")
        return A



# ===== VẼ ĐỒ THỊ Ở MỘT BƯỚC =====
def draw_step(adj_matrix, nodes, color_of, current_node, step):
    G = nx.Graph()
    G.add_nodes_from(nodes)
    n = adj_matrix.shape[0]

    for i in range(n):
        for j in range(i+1, n):
            if adj_matrix[i, j] == 1:
                G.add_edge(nodes[i], nodes[j])

    default_color = "lightgray"
    node_colors = [color_of.get(v, default_color) for v in nodes]

    plt.figure(figsize=(2, 2))
    pos = nx.spring_layout(G, seed=42)

    nx.draw(
        G, pos,
        with_labels=True,
        node_color=node_colors,
        node_size=300,
        font_color="white",
        font_weight="bold",
        edge_color="black"
    )

    nx.draw_networkx_nodes(
        G, pos,
        nodelist=[current_node],
        node_size=350,
        node_color=node_colors[nodes.index(current_node)],
        edgecolors="red",
        linewidths=3
    )

    plt.title(f"Bước {step}: tô đỉnh {current_node}")
    plt.axis("off")
    plt.show()



# ===== THUẬT TOÁN TÔ MÀU =====
def greedy_coloring_with_inline_images(adj_matrix):
    n = adj_matrix.shape[0]
    alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    nodes = list(alpha[:n])

    degree = [int(np.sum(adj_matrix[i])) for i in range(n)]

    print("Bậc từng đỉnh:")
    for i, d in enumerate(degree):
        print(f"  {nodes[i]}: {d}")
    print("-" * 40)

    idx_sorted = sorted(range(n), key=lambda i: degree[i], reverse=True)
    sorted_nodes = [nodes[i] for i in idx_sorted]

    print("Thứ tự duyệt:", " → ".join(sorted_nodes))
    print("=" * 40)

    all_colors = ["blue", "red", "yellow", "green", "orange", "purple", "brown"]
    color_of = {}
    step = 1

    for u in sorted_nodes:
        i = nodes.index(u)

        forbidden = set()
        for j in range(n):
            if adj_matrix[i, j] == 1:
                v = nodes[j]
                if v in color_of:
                    forbidden.add(color_of[v])

        chosen = next(c for c in all_colors if c not in forbidden)
        color_of[u] = chosen

        print(f"Bước {step}: xét đỉnh {u}")
        print("  Đỉnh kề đã tô:")

        for j in range(n):
            if adj_matrix[i, j] == 1:
                v = nodes[j]
                if v in color_of and v != u:
                    print(f"    {v}: {color_of[v]}")

        print(f"  Màu cấm: {sorted(forbidden)}")
        print(f"  → Gán màu cho {u}: {chosen}")

        draw_step(adj_matrix, nodes, color_of, u, step)

        print("-" * 40)
        step += 1

    return nodes, color_of, degree



# ====== CHẠY CHƯƠNG TRÌNH ======
print("=" * 40)

A = generate_constrained_random_adj_matrix(
    n=7,
    p=0.45,
    save_path="graph.txt"   # <== LƯU FILE TỰ ĐỘNG
)

print("Ma trận kề random hợp lệ (đã lưu vào graph.txt):")
print(A)
print("=" * 40)

nodes, color_of, degree = greedy_coloring_with_inline_images(A)

print("\nKẾT QUẢ CUỐI CÙNG:")
for i, v in enumerate(nodes):
    print(f"  Đỉnh {v} (bậc {degree[i]}) → {color_of[v]}")
```

```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# ---------- CẤU HÌNH ----------
np.random.seed(42)       # đặt seed để tái lập (bỏ nếu muốn hoàn toàn ngẫu nhiên)
SAVE_PATH = "random_graph.txt"
NODE_LABEL_START = ord('A')

# ---------- HÀM SINH MA TRẬN (RÀNG BUỘC) VÀ LƯU ----------
def generate_and_save_graph(n, p=0.45, save_path=SAVE_PATH):
    """
    Sinh ma trận kề sao cho mọi đỉnh có bậc >= 2 và không phải đường thẳng.
    Lưu vào save_path (ghi đè).
    Trả về ma trận A (numpy array)
    """
    attempts = 0
    while True:
        attempts += 1
        A = np.zeros((n, n), dtype=int)
        for i in range(n):
            for j in range(i+1, n):
                if np.random.rand() < p:
                    A[i, j] = 1
                    A[j, i] = 1
        deg = A.sum(axis=1).astype(int)
        if np.any(deg < 2):
            continue
        # kiểm tra path graph (đường thẳng) dạng: đúng 2 đỉnh bậc 1, còn lại bậc 2
        if (deg == 1).sum() == 2 and (deg == 2).sum() == n - 2:
            continue
        np.savetxt(save_path, A, fmt="%d", delimiter=" ")
        print(f"✔ Đã sinh và lưu ma trận vào '{save_path}' sau {attempts} lần thử.")
        return A

# ---------- HÀM VẼ ĐỒ THỊ TỪ MA TRẬN ----------
def plot_graph(A, nodes, color_of, highlight=None, title=None, figsize=(5,4)):
    G = nx.Graph()
    G.add_nodes_from(nodes)
    n = A.shape[0]
    for i in range(n):
        for j in range(i+1, n):
            if A[i, j] == 1:
                G.add_edge(nodes[i], nodes[j])
    # màu mặc định
    default_color = "lightgray"
    node_colors = [color_of.get(v, default_color) for v in nodes]
    pos = nx.spring_layout(G, seed=42)  # cố định layout cho mọi bước

    plt.figure(figsize=figsize)
    nx.draw(G, pos, with_labels=True, node_color=node_colors,
            node_size=500, font_weight='bold', font_color='white', edge_color='black')
    if highlight is not None:
        nx.draw_networkx_nodes(G, pos, nodelist=[highlight], node_size=650,
                               edgecolors='red', linewidths=3,
                               node_color=[color_of.get(highlight, default_color)])
    if title:
        plt.title(title)
    plt.axis('off')
    plt.show()

# ---------- CHƯƠNG TRÌNH TÔ MÀU (GREEDY + dynamic degree update) ----------
def greedy_coloring_trace(A, show_plots=True, pause_sec=0.8):
    n = A.shape[0]
    nodes = [chr(NODE_LABEL_START + i) for i in range(n)]
    degree = A.sum(axis=1).astype(int).tolist()
    original_degree = degree.copy()

    # thứ tự duyệt: sắp theo bậc ban đầu giảm dần (như yêu cầu trước)
    order_idx = sorted(range(n), key=lambda i: original_degree[i], reverse=True)
    order = [nodes[i] for i in order_idx]

    colors = ["Red", "Green", "Blue", "Yellow", "Orange", "Purple", "Brown", "Cyan", "Magenta"]
    color_of = {}
    used = set()

    trace_rows = []
    step = 1

    # hiển thị ma trận ban đầu & bậc ban đầu
    print("\nMa trận kề (A):")
    print(A)
    print("\nBậc ban đầu:")
    for i,v in enumerate(nodes):
        print(f"  {v}: {original_degree[i]}")
    print("\n--- Bắt đầu quá trình tô màu ---\n")

    # dùng 1 vị trí layout cố định để plot nhiều lần
    # lặp theo order đã tính
    for u in order:
        i = nodes.index(u)
        # tính màu cấm từ các láng giềng đã tô
        forbidden = set()
        for j in range(n):
            if A[i, j] == 1:
                v = nodes[j]
                if v in color_of:
                    forbidden.add(color_of[v])

        chosen = next(c for c in colors if c not in forbidden)
        color_of[u] = chosen
        used.add(u)

        # ghi dòng trace: lưu "Bậc khi chọn" là bậc hiện tại trước khi cập nhật
        row = {
            "Bước": step,
            "Đỉnh chọn": u,
            "Bậc khi chọn": degree[i],
            "Màu tô": chosen,
            "Màu cấm": ", ".join(sorted(forbidden)) if forbidden else "Không"
        }

        # cập nhật bậc động: giảm 1 cho các láng giềng chưa bị loại
        for j in range(n):
            if A[i, j] == 1:
                degree[j] -= 1

        # đỉnh u được loại (bậc -> 0)
        degree[i] = 0

        # thêm cột bậc cho mọi đỉnh (sau khi cập nhật)
        for v in nodes:
            row[f"Bậc {v}"] = degree[nodes.index(v)]

        trace_rows.append(row)

        # In chi tiết bước hiện tại ra console
        # (Bạn có thể format lại cho đẹp hơn nếu cần)
        print(f"Bước {step}: chọn {u} | Bậc khi chọn = {row['Bậc khi chọn']} | Màu tô = {chosen} | Màu cấm = {row['Màu cấm']}")
        print("  Bậc hiện tại:", "  ".join([f"{nodes[k]}={row[f'Bậc {nodes[k]}']}" for k in range(n)]))
        print("-"*60)

        # VẼ ĐỒ THỊ TẠI BƯỚC NÀY (nếu bật)
        if show_plots:
            title = f"Bước {step}: tô {u} = {chosen} (bậc trước={row['Bậc khi chọn']})"
            plot_graph(A, nodes, color_of, highlight=u, title=title, figsize=(4,3))
            # nhỏ pause để người dùng thấy thay đổi (trong notebook)
            time.sleep(pause_sec)

        step += 1

    df = pd.DataFrame(trace_rows)
    return color_of, original_degree, df

# ---------- HÀM VẼ ĐỒ THỊ CUỐI CÙNG ----------
def show_final_graph(A, color_of, figsize=(6,5)):
    nodes = [chr(NODE_LABEL_START + i) for i in range(A.shape[0])]
    plot_graph(A, nodes, color_of, highlight=None, title="Đồ thị sau khi tô xong", figsize=figsize)


# ---------- MAIN: nhập, sinh, lưu, chạy, hiển thị ----------
def main():
    try:
        n = int(input("Nhập số đỉnh (gợi ý 5-10): "))
        if n < 2:
            print("Số đỉnh phải >= 2")
            return
    except:
        print("Giá trị không hợp lệ. Dừng.")
        return

    # sinh & lưu ma trận (vẫn đảm bảo ràng buộc)
    A = generate_and_save_graph(n=n, p=0.45, save_path=SAVE_PATH)

    # chạy greedy + trace + hiển thị từng bước
    color_of, original_deg, df_trace = greedy_coloring_trace(A, show_plots=True, pause_sec=0.6)

    # hiển thị bảng trace bằng pandas
    print("\n\n================== TRACE TABLE (DataFrame) ==================\n")
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', 2000)
    display(df_trace)

    # vẽ đồ thị cuối cùng
    show_final_graph(A, color_of, figsize=(3,3))

    # in kết quả cuối cùng
    print("\nKẾT QUẢ CUỐI CÙNG:")
    for node in sorted(color_of.keys()):
        idx = ord(node) - NODE_LABEL_START
        print(f"Đỉnh {node} (bậc ban đầu {int(A.sum(axis=1)[idx])}) -> {color_of[node]}")

    print(f"\nTổng số màu sử dụng: {len(set(color_of.values()))}")
    print(f"Ma trận đã lưu vào file: {SAVE_PATH}")

# Chạy
if __name__ == "__main__":
    main()
```
