<b>1.1. Giới thiệu ràng buộc</b><br>
&emsp;&emsp;Ràng buộc trong thuật toán được hiểu là một quan hệ áp đặt lên một tập các biến, nhằm quy định những giá trị nào của các biến là hợp lệ. Các ràng buộc này giúp thu hẹp không gian nghiệm và đảm bảo lời giải thỏa mãn các điều kiện đã đặt ra. Việc biểu diễn ràng buộc có thể thực hiện theo nhiều cách khác nhau, phổ biến nhất là dùng các biểu thức toán học hoặc logic để mô tả quan hệ giữa các biến, hoặc sử dụng bảng liệt kê tất cả các tổ hợp giá trị hợp lệ mà các biến có thể nhận. Nhờ có ràng buộc, thuật toán có thể kiểm tra, loại bỏ hoặc đánh giá các nghiệm tiềm năng một cách hiệu quả, từ đó tìm ra lời giải đúng và phù hợp với yêu cầu của bài toán.<br>
<b>1.2. Bài toán thỏa mãn ràng buộc</b><br>
&emsp;&emsp;Bài toán thỏa mãn ràng buộc (Constraint Satisfaction Problem – CSP) là loại bài toán trong đó mục tiêu là tìm một bộ giá trị gán cho các biến sao cho tất cả các ràng buộc được đáp ứng đồng thời. Một CSP gồm ba thành phần chính: tập biến, miền giá trị của mỗi biến, và tập ràng buộc chi phối các biến đó. Nhiệm vụ của thuật toán là xác định nghiệm thỏa mãn toàn bộ các điều kiện, hoặc kết luận rằng không tồn tại nghiệm hợp lệ. Các bài toán CSP thường xuất hiện trong lập lịch, phân công công việc, tô màu đồ thị, sudoku, tối ưu hóa tài nguyên và nhiều bài toán trí tuệ nhân tạo.<br>
<b>1.2.1. Kỹ thuật giải quyết ràng buộc</b><br>
&emsp;&emsp;CSP có thể được giải quyết bằng nhiều kỹ thuật khác nhau, bao gồm:<br>
&emsp;&emsp;- Quay lui: Một thuật toán tìm kiếm theo chiều sâu mà xây dựng dần các ứng viên cho các giải pháp và từ bỏ một ứng viên ngay khi xác định rằng nó không thể dẫn đến một giải pháp hợp lệ.<br>
&emsp;&emsp;- Lan truyền ràng buộc: Kỹ thuật này giảm không gian tìm kiếm bằng cách áp dụng các ràng buộc cục bộ, điều này có thể đơn giản hóa vấn đề trước khi áp dụng các thuật toán tìm kiếm.<br>
&emsp;&emsp;- Phương pháp Heuristic: Những phương pháp này sử dụng các quy tắc kinh nghiệm để hướng dẫn quá trình tìm kiếm, cải thiện hiệu quả.<br>
&emsp;&emsp;Các tiến bộ gần đây trong CSPs tận dụng các kỹ thuật học máy để nâng cao khả năng giải quyết vấn đề, cho phép xử lý các ràng buộc phức tạp một cách hiệu quả hơn. Ví dụ, việc tích hợp CSPs với mạng nơ-ron có thể dẫn đến các giải pháp sáng tạo trong các môi trường động.<br>
