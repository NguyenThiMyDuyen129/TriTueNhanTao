<b>1.1. Giới thiệu ràng buộc</b><br>
<b>1.1.1. Khái niệm ràng buộc</b><br>
&emsp;&emsp;Ràng buộc (CSP) trong thuật toán được hiểu là một quan hệ áp đặt lên một tập các biến, nhằm quy định những giá trị nào của các biến là hợp lệ. Các ràng buộc này giúp thu hẹp không gian nghiệm và đảm bảo lời giải thỏa mãn các điều kiện đã đặt ra. Việc biểu diễn ràng buộc có thể thực hiện theo nhiều cách khác nhau, phổ biến nhất là dùng các biểu thức toán học hoặc logic để mô tả quan hệ giữa các biến, hoặc sử dụng bảng liệt kê tất cả các tổ hợp giá trị hợp lệ mà các biến có thể nhận. Nhờ có ràng buộc, thuật toán có thể kiểm tra, loại bỏ hoặc đánh giá các nghiệm tiềm năng một cách hiệu quả, từ đó tìm ra lời giải đúng và phù hợp với yêu cầu của bài toán. Các thành phần chính của CSP bao gồm:<br>
&emsp;&emsp;- Biến: Đây là các yếu tố cần được gán giá trị.<br>
&emsp;&emsp;- Miền: Mỗi biến có một miền, đó là tập hợp các giá trị có thể mà nó có thể nhận.<br>
&emsp;&emsp;- Ràng buộc: Đây là các quy tắc xác định mối quan hệ giữa các biến, chỉ định các tổ hợp giá trị nào là hợp lệ.<br>
<b>1.1.2. Ràng buộc trong thuật toán</b><br>
&emsp;&emsp;Trong một bài toán, ràng buộc là những điều kiện bắt buộc mà các nghiệm phải tuân theo để được xem là hợp lệ. Các ràng buộc này có thể xuất phát từ bản chất của bài toán, quy định nghiệp vụ, hoặc giới hạn kỹ thuật. Chúng quyết định cách các biến được phép nhận giá trị và cách chúng quan hệ với nhau. Nhờ các ràng buộc, không gian nghiệm được thu hẹp lại, giúp quá trình tìm kiếm lời giải trở nên chính xác và hiệu quả hơn. Ràng buộc có thể mang nhiều dạng, chẳng hạn như các biểu thức toán học, logic, hoặc bảng liệt kê những giá trị và tổ hợp giá trị được phép sử dụng.<br>
<b>1.1.3. Kỹ thuật giải quyết ràng buộc</b><br>
&emsp;&emsp;CSP có thể được giải quyết bằng nhiều kỹ thuật khác nhau, bao gồm:<br>
&emsp;&emsp;- Quay lui: Một thuật toán tìm kiếm theo chiều sâu mà xây dựng dần các ứng viên cho các giải pháp và từ bỏ một ứng viên ngay khi xác định rằng nó không thể dẫn đến một giải pháp hợp lệ.<br>
&emsp;&emsp;- Lan truyền ràng buộc: Kỹ thuật này giảm không gian tìm kiếm bằng cách áp dụng các ràng buộc cục bộ, điều này có thể đơn giản hóa vấn đề trước khi áp dụng các thuật toán tìm kiếm.<br>
&emsp;&emsp;- Phương pháp Heuristic: Những phương pháp này sử dụng các quy tắc kinh nghiệm để hướng dẫn quá trình tìm kiếm, cải thiện hiệu quả.<br>
